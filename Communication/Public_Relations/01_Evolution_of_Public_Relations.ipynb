# The Evolution Of Public Relations

- Managing relationships and communication between an organization and the public.  
- A planned, long-term, and continuous effort to build and maintain mutual understanding between an organization and its publics.  
- The strategic management of communication between an organization and the public to influence perception, trust, reputation, and decisions.

##### a ) Audience
People receiving messages (Passive).  
Example:

    TV viewers watching a Starbucks ad.

##### b ) Stakeholders
People financially or structurally affected.  
Example:

    investors
    employees
    suppliers
They have direct stakes.

##### c ) Public
Groups formed around shared interests/issues.  
Example:

    Environmental activists criticizing a factory.

They are publics because:

    they organize themselves
    share common concern
    influence public opinion

Modern PR heavily focuses on publics because:

    publics can destroy or strengthen reputation rapidly through social media.

Example:

    A TikTok boycott campaign against a brand.

## 1: Type & Functions
#### a ) Government
Purpose:

    Maintain legitimacy and public trust.

Example:

    Malaysia government explaining subsidy policies.

Advanced concept:

    Government PR often overlaps with:

        nation branding
        propaganda
        public diplomacy
        crisis communication

#### b ) Investor Relations
Purpose:

    Maintain investor confidence.

Example:

    A public listed company explaining quarterly losses.
    Bad investor PR can crash stock prices.

    When companies fail to communicate transparently, investors panic and sell shares. This directly affects market capitalization.

#### c ) Employee Relations
Internal PR.  
Purpose:

    Maintain employee morale and alignment.

Example:

    During layoffs, PR must communicate carefully internally.

Bad internal PR causes:

    low morale
    leaks
    union conflict
    reputational damage

#### d ) NGO PR
Purpose:

    Gain public support for social causes.

Example:

    Environmental NGOs creating campaigns against plastic pollution.

Their main weapon:

    emotional storytelling
    media framing
    public mobilization

## 2: Evolution of PR
### a ) Ancient PR (Babylonian/Sumerians)
messages on mud walls 3000 years ago.  
This proves:

    PR existed before modern media.

Why? because governments always needed:

    legitimacy
    obedience
    support
    influence
Even ancient rulers used communication strategically. This is early political PR.  


### b ) Modern PR
#### i ) Entertainment Publicity
Circuses used hype and sensationalism.  
This evolved into:

    celebrity PR
    entertainment marketing
    influencer culture

Modern equivalent:
    
    Movie PR campaigns.

Example:
    
    Marvel actors appearing in interviews before movie release.

Purpose:
    
    Generate anticipation and emotional engagement.

#### ii ) Corporate PR
Railroads and oil companies used PR because they were criticized heavily.

Why? Industrialization created:

    monopolies
    labor exploitation
    public anger
    Pollution

Corporations needed reputation management. This created:

    corporate communication
    crisis management
    press relations

    Modern ESG reputation management is an evolution of this


#### iii ) Evolution PR in Malaysia
Malaysia's PR evolution is strongly tied to colonialism, war, nation-building, racial unity & political stability. Unlike US PR which evolved through capitalism and business.

##### Phase 1: 1939
British established Information Department.  
Purpose: 
    
    Gain support during war.
    This was wartime propaganda PR.

Functions:

    control narratives
    distribute official information
    influence public opinion

Advanced concept:

    This is an early example of:
        Information warfare (Using information, media, communication, and narratives as a weapon to influence people's thinking, emotions, decisions, and behavior)

    Modern equivalent:
        Governments managing narratives on social media during crises.
 ##### phase 2: After WW2
British reputation collapsed after Japanese invasion.  
British were seen as "powerful rulers". But Japan defeated them quickly.  
This damaged:

    authority
    legitimacy
    public trust
So PR was used to restore confidence.

##### Psychological Strategy
The British communicated:

    "We were not defeated, only withdrew strategically"
    This is narrative reframing.

Modern equivalent:

    Companies reframing failures.

Example, Instead of:
    
    "Sales collapsed"

PR says:

    "The company is restructuring for long-term growth"
Same psychological principle.

##### May 13 Riot & PR Development
This is one of the most important PR turning points in Malaysia.  
The government realized:

    Poor communication can create national instability.

The report found:

    Lack of dissemination of government policies caused dissatisfaction.

Meaning:

    PR is not just promotion.

PR can affect:

    social cohesion
    racial harmony
    national security
    This is why ministries started appointing PR officers.

BERNAMA's Role acts as:

    national information distributor
    government news source
    agenda-setting institution

##### Advanced media theory, This relates to:

    Agenda Setting Theory

        Media may not tell people what to think,
        but tells people what to think about.

        If BERNAMA prioritizes certain issues,
        those issues become nationally important.

##### Mahathir Era (1980–2003)

This period modernized Malaysian PR massively. Why? Economic growth created:

    corporations
    globalization
    foreign investors
    privatization

All required sophisticated PR. As economies become more complex, communication complexity also increases. Because organizations now manage:

    investors
    regulators
    international media
    consumers
    activists
    employees

This creates demand for PR consultants:

    crisis managers
    corporate strategists
