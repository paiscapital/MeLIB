# Logic

## 1: What is Logic
Logic is the study of reasoning:

    How to form arguments
    How to evaluate whether arguments are correct or incorrect
    Logic studies methods and rules for correct thinking
    It evaluates whether an argument is:
    Valid (good reasoning)
    Invalid (bad reasoning)

Logic = thinking in a structured, correct, and systematic way\
<br>

## 2: Structure of an argument
Every argument has:

### a ) Premises
Premises are facts or statements you give as an evidence.  
They give reasons or support  

Premises = "What we already know"

### b ) Conclusion
The final statement derived from premises. The conclusion is the final idea you want to prove that it comes from the premises.  
Example:  

    Premise 1: All cats have tails
    Premise 2: Mimi is a cat
    Conclusion: Mimi has a tail

This is correct because premises support the conclusion.  
Conclusion = "What we are trying to show"  

An argument can be valid & logical structure is correct, but still false if premises are wrong.  
Example:

    Napoleon was Japanese ✗
    All Japanese are Asian ✔
    Napoleon was Asian

Structure is valid, but premise is false & not reliable.
<br>

Why Do We Study Logic?

You apply thinking skills to:

    Daily decisions

Example:

## 4: Types of Logical Method
### a ) Deductive 
Start with a general rule then apply it to a specific case. Start with theory then end with specific case.

    "If this rule is always true, then this specific result must also be true."
If all premises are true and the logic is correct, then the conclusion must be true (100% certain). No guessing, no probability and it is guaranteed.  

Example:

    Premise 1: All humans die
    Premise 2: A is a human
    Conclusion: A will die

The rule applies to all humans and A belongs to that group. So the conclusion is unavoidable.  

Deductive reasoning is:

    Certain with no doubt if premises are true
    Strict aand must follow logical structure
    Cannot make assumptions or guesses

It's like math, if 2 + 2 = 4, then it will always be 4.  

#### Top-Down Thinking
Deduction works from:

    Big Idea (general/theory) to Small Case (specific)

Example:

    General: All students must attend class
    Specific: Arif is a student
    Arif must attend class

That's why it's called top-down thinking.  

#### Flow
1: Theory

    A general rule or idea
    Example: "All humans will die"

2: Hypothesis

    Apply the rule to a case
    "If A is human, then A will die"

3: Observation

    Check the real situation
    "A is human"

4: Confirmation / Refutation

    If it matches then conclusion confirmed
    If not, reject it

### b ) Inductive
An Inductive approach is the process of drawing general conclusions based on specific observations.  
From specific to general. This means, you start with small observations (examples), then you make a big general conclusion.  
Example:

    Arif plays football
    Amin plays football
    Conclusion: All students play football

This can be wrong, because the problem is you only observed 2 students, not all students

Maybe:

    Sarah doesn't play football 
    John prefers basketball 

So the conclusion can be wrong.  

#### Flow
1: Observation 

    You see specific cases
    Arif plays football
    Amin plays football

2: Pattern 

    You notice a repeated behavior
    "Many students seem to play football"

3: Hypothesis 

    You make a possible explanation
    "Maybe students like football"

4: General Theory 

    You form a broad conclusion
    "All students play football"

### c ) Abductive 
Making a smart guess to explain a situation based on limited information, you don't have full proof but you choose the explanation that makes the most sense.  

    You observe something
    You ask: why did this happen?
    You choose the best possible explanation

Example:

    Aminah is smiling and holding her results
    Best explanation: She got good results

You didn't see her marks, But your brain connects the clues and makes a logical guess.  

#### Science
    Plants are dying
    Best explanation: Not enough water or sunlight

#### Investigation
    Floor is wet
    Best explanation: Someone spilled water or it rained

#### Diagnosis (Doctor)
    Patient has fever + cough
    Best explanation: Infection

#### Thinking Backwards
Abductive reasoning works backwards.  

    Start with result
    Then guess the cause

Example:

    Result: Aminah is happy
    Cause (guess): She did well

Abductive reasoning is not always correct, there can be many possible explanations.  
Example, aminah is smiling, possible reasons:

    Good results 
    Got a gift 
    Good news 

But we choose the most likely one.

### d ) Heuristic Method
A heuristic is a quick way or shortcut to solve a problem. Instead of thinking too deeply, you try simple methods or use experience to find an answer.  
    
    Exploration + trial and error.
    "Try different ways until something works"
    "Use your experience to make a fast decision"

Characteristics:

    Flexible thinking
    Practical problem-solving
    Not strictly logical

How it works:

    You don't follow strict rules
    You experiment or guess smartly
    You learn from mistakes

#### Example 
1; Studying  
You try:

    Reading notes
    Watching videos
    Drawing diagrams
You keep what works best, That's heuristic thinking.  

2: Daily Life
You can't open a jar. You try:

    Twist harder
    Use cloth
    Tap it

You don't know the exact method, you just try different ways.  

3: Decision Making
You choose food based on:
    
    "I liked this before"

That's using experience as a shortcut.  
Heuristic can sometimes be wrong, because It's based on shortcuts, not deep analysis.  


#### Trial & Errors
You try different ways until one works.  

    Try method A ✗
    Try method B ✗
    Try method C ✔ Success
Example, arif tries:

    Reading notes ✗
    Watching videos ✗
    Drawing diagrams ✔
He learns best with visual notes.  
Trial & Error = Learn by trying until you succeed.  

#### Lateral Thinking
Solve problems by changing your perspective. Instead of doing the same thing, you:

    Think differently
    Look for a smarter solution
Example, problem:

    Arif feels sleepy in class
    Normal thinking: Drink coffee 
    Lateral thinking: Fix sleep schedule 

Solves the real problem, not just the symptom.  
Lateral Thinking = Solve problems by thinking differently.  

#### Principles
1. Freedom: You are free to try any method (No strict rules).
2. Experience: Use what you learned before (Your past helps you decide).
3. Learning by Doing: You learn better by actually doing, not just reading.
4. Usefulness: Choose methods that work in real life.
5. Logical Thinking: Even though it's flexible, still think rationally.
6. Enjoyment: Learning is better when it's fun and interesting.
7. Individual Effort: You must try and explore yourself.

## 5: Evaluating Argument
When we evaluate an argument, we are asking "Is this argument good or bad?". To answer that, we look at two things:

    The structure (logic)
    The truth of the premises (facts)

### a ) Deductive Arguments
Focus on certainty (100% true if correct)
Valid meaning:

    The structure is correct
    If the premises are true, the conclusion must be true
Example:
    
    All humans are mortal
    Ali is a human
    Ali is mortal

This is valid because the logic is perfect.  
Invalid meaning:

    The structure is wrong
    Even if premises are true then conclusion may be wrong
Example:
    
    All cats are animals
    A dog is an animal
    A dog is a cat

Wrong structure, invalid argument

### b ) Non-Deductive Arguments (Inductive, Abductive)
Focus on probability (not 100%).  
Strong meaning:

    Premises give good support
    Conclusion is likely true
Example:
    
    90% of students pass this exam
    Ahmad studied hard
    Ahmad will probably pass

Strong, but not guaranteed.  
Weak meaning:

    Premises give little or no support
Example:
    
    Ahmad likes football
    Ahmad will pass the exam

No connection & weak argument.

### c ) Truth of Premises
Are the premises true or false?
#### True Premises
Make argument more reliable.  
Example:

    All humans need oxygen ✔
    Ali is human ✔
    Ali needs oxygen ✔

👉 Strong and true, very good argument

#### False Premises
Make argument unreliable (even if structure looks correct).  
Example:

    All birds can swim ✗
    Eagle is a bird ✔
    Eagle can swim

👉 Structure may be valid, but premise is false & argument is weak overall.

## 6: Fallacies
A fallacy is a mistake in thinking or reasoning.

    You are trying to argue something
    But your thinking has an error
    So your argument becomes weak or wrong
The conclusion may sound right, but the way you got there is wrong.

### a ) Formal Fallacies
The logic structure is wrong, The argument is built incorrectly. 
The "formula" of thinking is broken.  
Example:

    All cats are animals
    All dogs are animals
    Therefore, dogs are cats  ✗
Problem:

    Structure is wrong
    Just because both are animals doesn’t mean they are the same

### b ) Informal Fallacy
The idea, meaning, or reasoning is wrong, The structure may look okay, but the logic inside is weak.  
The "content" of thinking is flawed.  

### Type of Informal Fallacies
#### a ) Ad Hominem (Personal Attack)
Instead of attacking the argument, you attack the person.  
Example:

    "Arif is wrong about cleanliness because he is dirty"

Problem:

    You are ignoring the argument
    You are focusing on the person
Correct thinking:

    Judge the idea, not the person

#### b ) Begging the Question (Circular Reasoning)
You assume something is true without proving it. You use the conclusion as the reason.  
Example:

    "Philosophy is useless, so I won't study it"

Problem: 

    It assumes philosophy is useless
    But gives no proof
Correct thinking:

    Always give evidence, not assumptions

#### c ) False Generalization (Hasty Generalization)
You make a big conclusion from small or weak evidence.  
Example:

    "Tiger and lion are dangerous killers, so cats are also dangerous killers"

Problem:

    Tigers/lions != all cats
    You generalized too quickly
Correct thinking:

    Use enough evidence before making general conclusions
