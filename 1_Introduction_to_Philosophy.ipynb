# Introduction to Philosophy

## 1: What is philosophy
Philosophy = "love of wisdom"  +
From Greek:
    
    philo = love
    sophia = wisdom

Philosophy is the effort to understand truth, life, and reality (A way of thinking).

    Asking questions (especially "Why?")
    Thinking critically
    Analyzing ideas
    Challenging assumptions
Example:  
Someone says: "Only Muar has authentic mee bandung”  
Philosophy asks:

    What does "authentic" mean?
    Who decides that?

Philosophy is called "The mother of knowledge", Because it connects to Science, Religion, Mathematics, Psychology & Technology.  

### Key thinker
Philosophy developed in 3 main regions (around 600 BC–300 BC):  
    
    West (Greek)
    Thinkers: Socrates, Plato, Aristotle
    Focus : logic, ethics, knowledge

    
    India
    Systems: Veda, Buddhism, Jainism
    Focus: life, suffering, spirituality

    
    China
    Thinkers: Confucius, Laozi
    Focus: morality, society, harmony
    
#### Socrates
Famous idea: "I know that I know nothing"  
Meaning:
    
    A wise person knows their limits
    Philosophy = not knowing everything, but seeking truth

### Philosophy = Investigating Ideas
Philosophy is like a "thinking tool" that helps you not accept things blindly, but instead investigate them carefully.  

#### a ) Check Information
Before you believe something, you verify it first. Ask questions like:

    Where did this come from?
    Is it reliable?
    Is there evidence?
Example, you see a viral post:
    
    "Students were bullied in a university"

Don't believe immediately & check:

    Original source
    Full video (not edited clips)
    Official statements

#### b ) Avoid Blind Belief
Don't believe something just because:

    Many people say it
    It's emotional or shocking
    It matches your opinion
Philosophy teaches:

    "Just because something looks true, doesn't mean it is true."

Example:

    Viral video looks real
    People are angry in comments

Blind belief: "It must be true!"  
Philosophical thinking: "Let me verify first."  

#### c ) Think Deeply 
Don't just look at what is happening, ask why and how.  
Go deeper, ask:

    What really happened?
    Is there missing context?
    Who benefits from this information?
Example, instead of just:

    "This video is bad"
Think:

    Is it edited?
    Is it taken out of context?
    Is it meant to provoke emotions?

## 2: Philosophy in different cultures
All cultures try to answer the same question:

    "What is truth and how should we live?”

Same goal: truth & good life  
Different methods: different ways of thinking  

### a ) Greek Philosophy
Use logic and rational thinking to find truth. 

    Ask questions (Socratic method)  
    Use arguments and logic  
    Don't just accept beliefs  

Example:  
Instead of saying "This is right because people say so".  
They ask:

    Why is it right?
    Is there evidence?
Focus:

    Truth through reason (akal)
    Clear thinking

### b ) Chinese Philosophy
Live in balance and harmony.  
Main traditions:

    Confucianism → social harmony
    Taoism → harmony with nature
    
How they think:
    
    Focus on relationships (family, society)
    Emphasize balance (yin & yang)

Example:

    If there is conflict:
        The goal is not to "win"
        But to restore harmony

Focus:

    Good behavior
    Peaceful society
    Balance in life

### Indian Philosophy
Escape suffering and reach inner freedom  
Concept:

    Life has suffering
    Goal = moksha (liberation)
How they think:

    Look inward (self-reflection, meditation)
    Focus on soul, karma, rebirth

Example:

    Instead of asking:
        "Who is right?"
They ask:

    Who am I?
    How do I free myself from suffering?

Focus:

    Spiritual truth
    Inner peace
    Freedom from cycle of life

### Islamic Philosophy
Combine akal (reason) + wahyu (revelation).  
Famous thinkers:

    Al-Kindi
    Ibn Sina
    Ibn Rushd

How they think:

    Use logic AND religion together
    Philosophy supports faith

Example:
    
    Study nature → understand God better

Focus:

    Truth from both reason and faith
    Harmony between science, philosophy, religion

## 3: Difference
These three are different ways humans try to understand truth:

    Religion → believe  
    Science → test  
    Philosophy → think  

### a ) Religion (Faith-Based)
Focus:

    God / higher power
    Meaning of life
    Moral rules (what is right/wrong)
    Life after death
    
Method:

    Faith (iman) → believe without needing proof
    Revelation → teachings from holy books or prophets
    Tradition & authority
    
Example, question: "What happens after we die?"  
Religion answer:

    There is heaven/hell
    Based on religious teachings
Religion gives certainty and guidance, even for things we cannot test.

### b ) Science (Evidence-Based)
Focus:

    Physical world (nature, body, universe)
    Things you can see, measure, test

Method:

    Observation 
    Experiment  
    Data & evidence 
    
Example, question: "Why does it rain?"  
Science answer:

    Water evaporates → forms clouds → falls as rain

Science explains how things happen, using proof.  

### c ) Philosophy (Thinking-Based)
Focus, big questions:

    What is truth?
    What is reality?
    What is a good life?

Method:

    Logic
    Reasoning
    Critical thinking
Example, question: "What is a good life?"  
Philosophy answer:

    Depends on happiness, virtue, purpose
    No single final answer → requires thinking

### d ) Conclusion
Same Question, Different Answers.  
Question: "Why do humans exist?"  

Religion:

    Created by God for a purpose
Science:
    
    Result of evolution
Philosophy:
    
    What does "purpose" even mean?

| Aspect | Religion | Science | Philosphy |
|--------|----------|---------|-----------|
| Source | Faith | Evidence  | Reason |
| Method | Believe | Test | Think |
| Focus | Meaning & God | Physical World | Deep Questions |
| Certainty | High (Belief) | Temporary (Can Change) | Open-Ended |

## 4: Religion & Philosophy
Reason = Using your mind to think logically (Philosophy).  
Faith = Believing without needing proof (Religion).
  
#### We Need Both (Balance)
Many philosophers say:

    Reason helps us understand the world
    Faith gives meaning and purpose
Example:

    Science explains how life works
    Faith explains why life matters

The idea is they complete each other.

#### Only Reason is Enough
Some thinkers say:

    Only evidence = truth
    Faith can lead to mistakes

Example:
    
    Believing something without proof can be wrong

The idea is trust logic only

#### Faith is More Important
Some believe:

    Human reason is limited
    Not everything can be proven

Example:

    Love
    Meaning of life
    God

Faith gives higher understanding

### a ) Harmony (Work Together)
Religion and philosophy support each other, not fight.

    Religion gives beliefs (faith, God, meaning)
    Philosophy gives reasoning (logic, explanation)

Together = stronger understanding  
Example:

    Religion says: "God exists"
    Philosophy asks: "Can we prove or explain this logically?"

Philosophy helps:

    Explain religious ideas
    Defend beliefs
    Solve ethical problems
Example thinker (Ibn Rushd):

    Said religion and philosophy both aim for truth
    No real contradiction between them

    Faith + Reason = Complete understanding

### b ) Conflict
Religion and philosophy are in tension or disagreement.

    Religion = based on faith
    Philosophy = based on questioning

The problem is philosophy may challenge religious beliefs.  
Example:

    Religion: "Believe in God"
    Philosophy: "What evidence is there?"

Conflict happens when:

    Religion says "don't question"
    Philosophy says "question everything"
Example thinker (Tertullian):

    Believed faith is more important than reason
    Questioned philosophy's role in religion

    Faith vs Reason = Tension

### c ) Subordination (Philosophy Follows Religion)
Philosophy is less important and must follow religion.  

    Religion = main authority (truth comes first)
    Philosophy = tool to support religion

Philosophy cannot go against religious teachings.  
Example:

    Religion: "This is true"
    Philosophy: "Let me explain why it is true"

Philosophy's role:

    Explain beliefs
    Defend religion
    Not challenge it

Limitation:

    Critical thinking is restricted
    Cannot question core beliefs

    Religion first, philosophy supports

## 5: Branch of Philosophy
Each branch studies a different type of question.

### a ) Metaphysics 
This is about what truly exists (what is reality).  
Metaphysics = trying to understand the nature of existence  
It asks:

    What is real?
    Does God exist?
    Do we have free will?
    What is the nature of the universe?

Example:

    Are dreams real or just in your mind?
    Is time real, or just something we created?

### b ) Epistemology
This is about knowledge.  
Epistemology = checking if what we know is really true  
It asks:

    What is knowledge?
    How do we know something is true?
    Can we trust our senses?
    What is the difference between belief and knowledge?

Example:

    You see smoke & you believe there is fire
    But how sure are you?

### c ) Logic
This is about good reasoning.  
Logic = rules for clear and correct thinking   
It teaches:

    How to avoid mistakes in thinking
    How to build strong arguments

Example, wrong logic:
    
    "All gamers are lazy. Ali is a gamer → Ali is lazy"

Problem: assumption, not proof

### d ) Ethics
This is about morality.  
Ethics = guide for how to behave properly  
It asks:

    What is good or bad?
    What should we do?
    What is a good life?

Examples:

    Is lying always wrong?
    Is stealing okay if you are starving?

#### Different Ethical Views
Absolute (Rule-Based Ethics):

    Stealing is always wrong, no exceptions.
Example, Immanuel Kant:

    Believed moral rules must be followed no matter what
    So even if starving, stealing is still wrong
Morality = strict rules


Consequences-Based (Flexible):

    Judge actions by the outcome

Example:

    If stealing food saves a life, it may be justified
    Saving a life is more important than property

### e ) Political
This is about society, power, and government.  
Political philosophy = how to organize a fair society  
It asks:

    What is justice?
    What is the best government?
    What rights do people have?
Example:

    Should everyone have equal wealth?
    Is democracy the best system?

### f ) Aesthetics
This is about art and beauty (what is beauty).  
Aesthetics = understanding beauty, art, and taste   
It asks:

    What makes something beautiful?
    What is art?
    Is beauty objective or subjective?

Example:

    Why do some people love a painting, others don't?
    Is music beautiful for everyone?

## 6: Philosophy in Daily life
You're stuck in a traffic jam and you start feeling stressed or angry. Philosophy helps you think better, calmer, and smarter using 3 tools:

### a ) Logic
What people usually do:

    "These drivers are stupid!"
    "People here are so bad!"

The problem is you are assuming without evidence. Philosophy (logic) says:

    "Do I have proof?"
    "Am I jumping to conclusions?"

Better thinking:

    "I don't know the real reason yet."
    Don't assume, think first

### b ) Reality
Now you look at facts instead of emotions, possible reasons:

    Accident 
    Road construction 
    Rain or weather 
    Traffic light problem 

Philosophy teaches, not everything is under your control (external factors).   
The world is complex, not everything is someone's fault.  
<br>

### c ) Reality
Now you decide how to behave:

    Bad reactions:

        Honking angrily
        Shouting or insulting
        Getting aggressive

    Philosophical (ethical) response:

        Stay calm
        Be patient
        Respect others
        Inform others if you're late

"Even if the situation is bad, I can choose to act well."

## 7: Purpose
Knowledge = You understand truth.   
Self-improvement = You become wiser.   
Change the world = You apply it to society.  

### a ) Gain Knowledge 
Philosophy helps us understand what is true and real.

    Asking: What is truth?
    Questioning: Is what I believe actually correct?
    Thinking beyond surface-level answers
Example, you hear: "Money brings happiness"  
Philosophy asks:

    Is that always true?
    What is "happiness"?
    Can poor people be happy?

You don't just accept ideas, you analyze them. Philosophy trains your mind to think deeply and not be easily fooled.  

### b ) Improve Yourself
Philosophy helps you grow as a human being.  

    Building good character
    Controlling emotions
    Making better decisions
    Living a meaningful life
Example, you feel angry at someone. Philosophy asks:

    Is my reaction right or wrong?
    What is the “good” action here?
    Should I forgive or fight?

It guides your behavior.  
Philosophy = "How should I live?"  


### c ) Change the world (Social Impact)
Philosophy is not just thinking, it can change society.   
Example thinker (Karl Marx), He believed philosophy should not just explain the world  
It should change it:  

    Question injustice
    Challenge unfair systems
    Improve society
Real-life examples:
    
    Fighting corruption
    Speaking against inequality
    Promoting justice and rights

    Philosophy can turn to action


Philosophy can turn ideas into real-world change
