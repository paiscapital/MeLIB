# Theories & Models SBCC

# Individual Level
Even if people know something is bad, they still don't change. Example, people know smoking is dangerous but still smoke, because behaviour depends on:

    beliefs
    emotions
    confidence
    social pressure
    environment

All theories answer this question:

    "Why do people not change, and how can we make them change?"

Each theory focuses on a different factor.
| Theory | Focus                                    |
| ------ | ---------------------------------------- |
| HBM    | Personal beliefs about risk              |
| TPB    | Intention + social pressure              |
| TTM    | Stages (step-by-step change)             |
| SCT    | Environment + learning from others       |
| PMT    | Fear + coping ability                    |
| FBM    | Simplicity (motivation + ease + trigger) |
<br>

## 1: Health Belief Model (HBM)
"People act based on what they BELIEVE about risk"

#### Key parts 
    Susceptibility → "Can this happen to me?"
    Severity → "How serious is it?"
    Benefits → "Does this help me?"
    Barriers → "What stops me?"
    Cues to action → "What reminds me?"
    Self-efficacy → "Can I actually do it?"

Example (Vaccination):

    "I might get flu" → susceptibility
    "Flu is dangerous" → severity
    "Vaccine protects me" → benefit
    "Needles are scary" → barrier
    "Doctor reminds me" → cue
    "I believe I can go" → self-efficacy

If all are positive, behaviour happens. Self-efficacy is the strongest factor,
Benefits alone sometimes don't work anymore. People already know benefits, but still don't act, so we must:

    increase confidence
    reduce barriers

Weakness:
    
    Assumes people are rational
    Ignores emotions, habits, culture

## 2: Theory of Planned Behaviour (TPB)
Behaviour = Intention (niat)

#### 3 Key Drivers of Intention
    Attitude → "Do I like this behaviour?"
    Subjective Norm → "Do people expect me to do it?"
    Perceived Control → "Is it easy or hard?"

Example (Exercise):

    "Exercise is good" → attitude
    "My friends go gym" → norm
    "Gym is near me" → control

Strong intention, higher chance of action. Social pressure is very powerful in some cultures:

    Community > individual

So:
    
    Use influencers, family, peers
    Not just personal motivation

## 3: Transtheoritical Model (TTM)
Change happens in stages, not instantly.

#### 5 Stages
    Precontemplation → Not thinking to change
    Contemplation → Thinking about it
    Preparation → Ready soon
    Action → Doing it
    Maintenance → Keep doing

Example (Smoking):

    "Smoking is fine" → precontemplation
    "Maybe I should quit" → contemplation
    "I'll quit next month" → preparation
    "I stopped" → action
    "Still not smoking" → maintenance

80% people are not ready to act, So:

    Saying "Quit now!" doesn't work

Instead:

    Early stage: awareness
    Later stage: action strategies

##### Common Mistake
Many campaigns:

    Give same message to everyone

But TTM says:

    Message must match stage
<br>

## 4: Social Cognitive Theory
People learn from Environment + Others. Core idea is reciprocal determinism

#### 3 things influence each other

    Person (beliefs)
    Behaviour
    Environment
    
1. Observational Learning

        "I copy others"

Example:
    
    Seeing influencer exercise, you follow

2. Self-efficacy

        "I believe I can do it"

3. Outcome Expectations

        "What will happen if I do this?"

From studies:
    
    Self-efficacy = strongest predictor again

So:

    Show role models
    Use peer influence
    Build confidence step-by-step

## 5: Protection Motivation Theory (PMT)
Fear works only if people feel capable

#### Two Systems

1. Threat Appraisal (How dangerous? Can it happen to me?)

2. Coping Appraisal (Does solution work? Can I do it? Is it costly?)

Example (Anti-smoking ad):

    "Smoking causes cancer" → threat
    "You can quit easily" → coping
    
If fear high but coping low:

    People deny
    Ignore message

So:

    Fear + solution must go together

## 6: Fogg Behaviour Model (FBM)
Behaviour is simple.  
Formula:

    Behaviour = Motivation + Ability + Prompt

Example (Drinking water):

    Motivation → "I want to be healthy"
    Ability → water is easy to access
    Prompt → reminder notification

All 3 = behaviour happens
Easiest way to change behaviour:

    Not increase motivation
    Reduce difficulty (ability)

Example:

    Put healthy food in front
    Make unhealthy harder

<br>

Behaviour change is not simple or instant. It is a complex interaction of:

    beliefs
    social influence
    emotional responses
    confidence
    environmental factors

Effective SBCC strategies must:
    
    build self-efficacy
    reduce barriers
    match messages to readiness stage
    use social influence
    balance fear with solutions
    make behaviours easy and well-timed

# Society & Community Level

People don't act alone, behaviour is shaped by social norms, culture, and environment. Even if a person wants to change, they might NOT change because:

    Family disagrees
    Society judges
    No support
    Environment blocks them
Individual level = "inside the person"
Community level = "outside the person"
<br>

## 1: Social Norms Theory 
People follow what they think others do.

#### 3 Types of Norms
1. Descriptive Norm

        "What people do"
    Example:
        "Most people smoke"
   
2. Injunctive Norm

        "What people approve"

   Example:

   
        "My family thinks smoking is okay"

   
3. Collective Norm

        "Reality (actual data)"

    Example:

        Actually only 30% smoke

People often have wrong perceptions:

    Think bad behaviour is common
    Think good behaviour is rare

#### Powerful Strategy
Correct the perception:

    "Most people don't smoke"
    Then behaviour changes automatically

#### Big Mistake (Boomerang Effect)
Saying "80% people do bad behaviour" can make it worse, Because it normalizes the bad behaviour.

<br>

## 2: Diffusion of Inovations
How new ideas spread in society. Behaviour spreads like:

    Fire
    Viral trend

#### 5 Types of People

    Innovators → first try
    Early adopters → influencers
    Early majority
    Late majority
    Laggards → last
You don't need everyone, you just need influencers (early adopters), Then others follow.  
#### 5 Factors for Adoption
    Advantage → is it better?
    Compatibility → fits culture?
    Complexity → easy or hard?
    Trial → can try first?
    Visibility → can others see it?

Adoption is not instant, It follows an S-curve:

    Slow → Fast → Slow

<br> 

## 3: Community Mobilisation
Change is done by the community, not to them

#### Core Idea

Old way:

    Experts tell people what to do

New way:

    Community decides + acts

Process:
    
    Build trust
    Identify problems
    Plan together
    Act together
    Sustain change

Real change happens when:

    People feel ownership
    People work together

FGM didn't stop until many villages agreed together, Because social pressure must change at group level.

<br>

## 4: Participatory Communication 
Communication = dialogue, not lecture.

Old Model:
    
    Government → People (one-way)

Participatory:
    
    People ↔ Government (two-way)

People are not passive, they are partners.

#### Two Types of Empowerment
1. Cognitive (Understand problems deeply)
2. Relational (Gain power to act + negotiate)

Fake participation (tokenism): People are included only on the surface, but they don't have real power or influence.

#### Fake Participation
    Community meeting is held
    People share ideas
    But organisation ignores them
    Plan is unchanged
People are just for "show"

#### Real Participation
    Community gives ideas
    Their ideas are USED
    They help decide actions
    They lead implementation
That is true empowerment

<br> 

## 5: Social-Ecological Model (SEM)
Behaviour comes from many layers

#### 5 Levels
    
    Individual
    Interpersonal (family, friends)
    Community
    Organisation
    Policy

Example (Obesity):

    Individual → knowledge
    Family → eating habits
    School → food options
    Community → parks
    Policy → food laws
You cannot fix behaviour at one level only  
Just education = FAIL  
Must change environment too  

<br> 

## 6: Social Capital Theory 
Relationships = power  
Social networks = resources

#### Two Types
1. Individual Social Capital (Support from friends/family)
2. Collective Social Capital (Trust + cooperation in community)

Example, Strong community:

    People help each other
    Share information
    Act together

Weak community:

    No trust
    No change

Behaviour change spreads faster in High-trust communities.  
<br>

Behaviour change at the community level requires transforming not only individual beliefs but also the social norms, cultural expectations, and structural environments that shape behaviour. Theories such as Social Norms Theory and Diffusion of Innovations highlight the role of peer influence and social networks, while Community Mobilisation and Participatory Communication emphasise empowerment and collective action. The Social Ecological Model and Social Capital Theory further demonstrate that sustainable change occurs when interventions address multiple levels of influence and leverage strong community relationships.
