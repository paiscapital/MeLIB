# Health Belief Model

a psychological model used to explain and predict why people take (or don't take) health actions. It focuses on people's perceptions (beliefs) rather than just knowledge.

    HBM is used to understand behavior in prevention, screening (Filter), and treatment

Perceive = felt/feel
## Perceived Susceptibility (Risk)
It works because humans don't act based on actual risk, but on perceived (felt) risk.

    "Can this happen to me?"
    Smokers: "Others get cancer, I won't"
    Young people: "I'm too young for diabetes"
    
If high, more likely to act.
If low, ignore behavior, no lifestyle change.

## Perceived Severity (Seriousness)
It's not the actual danger, but the persons belief about how bad the outcome is. So two people can face the same disease, but:

    One thinks "it's nothing", no action
    One thinks "this will ruin my life", takes action

    "How serious is it?"
    Includes: death, disability, cost, social impact

"Cancer is deadly", motivates screening (filter).  

## Perceived Benefits
Human brain is goal-oriented (cost-benefit thinking). They believe the action will produce a valuable outcome.

    "Will this action help me?"

Must believe action is effective. Example, "Exercise reduces heart disease”

## Perceived Barriers

    "What's stopping me?"
    Cost, fear, stigma, time, access

Barriers often stronger than benefits in real behavior. Example, "I'm shy to do screening"

## Cue to Action
Even if a person feels at risk, knows it's serious & believes in benefits,
they often won't act until something "pushes" them.

    "What triggers me to act?"

    Internal: symptoms
    External: campaigns, reminders

Example, doctor advice, social media & illness of friend.

## Self-Efficacy (Confidence)
Self-efficacy is a person's belief in their ability to successfully perform a specific behavior under real-life conditions.

    "Can I actually do it?"

Added later to strengthen model

Example, "I know how to use a condom properly"

## HBM Structure
a ) Individual Perceptions (Susceptibility, Severity).  
b ) Modifying Factors (Age, culture, education, experience) and these shape perception.  
c ) Likelihood of Action (Benefits vs barriers, Cue to action & self-efficacy).  
  
This structure explains why two people behave differently even with same knowledge.  

## HBM in SBCC
In SBCC, we use HBM to design interventions, Instead of just giving information, we change beliefs strategically.

<img src="https://upload.wikimedia.org/wikipedia/commons/b/bc/Health_belief_model_diagram.jpg" alt="HBM" style="width: 550px; height: 300px; margin: auto;"><br>
| HBM Component | SBCC Strategy |
|---------------|---------------|
| Susceptibility | Show risk (statistics, stories) |
| Severity | Show consequences (emotional + real) |
| Benefits | Promote positive outcomes |
| Barriers | Reduce fear, cost, stigma |
| Cue to Action | Campaigns, reminders |
| Self-efficacy | Teach skills |

Example HIV Prevention:

    Condom use depends on:
        perceived severity of HIV
        perceived benefits
        self-efficacy (strongest factor)

SBCC intervention:

    Demonstrate condom use
    Reduce stigma
    Increase confidence

Real Application (HBM): 

    Cancer screening
    Smoking cessation
    HIV prevention
    Chronic disease (CKD)

People avoid screening due to embarrassment, inconvenience & fear.  
#### HBM Limitations 

1. Ignores social/environment factors (Focus on individual only)
2. Assumes people act rationally (Human are not always rational)
3. Doesn't include habits or emotions well (Focus only in cognitive not habits, emotions)
4. Hard to change long-term behavior (works in short term actions, strugle in lifestyle changes)

Example, even if someone knows smoking is harmful, addiction overrides belief.  

HBM is cognitive model, meaning its focus in thinking (beliefs).  
But behavior is also influenced by:

    Emotion
    Social norms
    Environment

That's why in SBCC, we often combine HBM with:
1. TTM (Stages of Change)
2. PMT (Protection Motivation Theory)
