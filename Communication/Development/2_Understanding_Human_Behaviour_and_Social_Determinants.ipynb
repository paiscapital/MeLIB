# Understanding Human Behaviour and Social Determinants in SBCC

## 1: Core SBCC
Engineering human behavior across systems.
Old Thinking (solution = give information)

    People behave badly because they don't know better.

New SBCC Thinking

    Human behavior = complex system, not just knowledge

Behavior is driven by:

    Mind (psychology)
    People (social pressure)
    System (environment / structure)

Information alone != behavior change.  

    Behavior = f(Individual + Social + Environment)  
    Behavior Change = (Psychology) × (Social Norms) × (Environment) × (System Design)
    
if any of these fails = behavior fails 
If any = 0 → whole system fails

#### Framework 
1. Diagnose:
   Where is the problem?
   (Mind / Social / System)
3. Identify bottleneck:
   What is blocking behavior?
4. Apply multi-level solution

## 2: Levels of Human Behavior
###  a ) Individual (Internal system)
    Knowledge → what you know
    Beliefs → what you think is true
    Attitudes → how you feel about it
    Self-efficacy → "Can I actually do this?"
    Risk perception → "Is this dangerous to me?"

Even if you know something is good, you want to do it but you still might not do it. Because you think you can't (low self-efficacy) or something blocks you externally (social/system).
<br>

### b ) Social (People around you)
This is group pressure system that include :

    Friends (peer)
    Family (peer)
    Community
    Culture
    
key Forces from this:

    Peer pressure = Smoking / Vaping:
        You personally know smoking is bad & Don't want to smoke. But your friends is all vape & say "just try once" & laugh if you refuse. You feel pressure, fear of being "left out" & fear of being "uncool". Result is you smoke (You didn't change your mind but you changed your behavior to fit the group).

    Social approval / rejection = Social Media: 
        You post something, if many likes/comments you feel good, but when no likes you feel bad. You start posting content that people like, not what you actually want (Behavior shifts toward approval optimization, not truth)

        
    Social norms = Classroom Behavior:
        In one class, veryone studies seriously but in another class everyone jokes and doesn't care (same student, same brain, different behavior). This because "this is how people behave here" (you don't decide behavior consciously, you copy the environment standard - Social cognitive Theory)

    Collective belief = Stock Market Hype:
        During a bull run people believe "This coin will go to the moon”, "Everyone is making money". Even if you are unsure, you start buying because "If everyone believes it, it must be true" (truth is replaced by shared belief)
        
    Collective Efficacy = Gym:
        Alone you feel lazy & you skip workouts, but in a group everyone trains together, encourages each other & pushes harder. You perform better than alone, because "We can do this together" (Group confidence > individual confidence)

### c ) Environmental (System)
This is real-world constraints:

    Infrastructure (clinics, roads)
    Money
    Government policy
    Media access
    Resources availability
You can't perform a behavior if the system makes it impossible, but most systems don’t fully prevent behavior, they only constrain it.
Example:

    Want to use contraception ✅
    Community supports ✅
    But clinic has no supply ❌
    Behavior = impossible

            1) Teach students about malaria prevention
            2) Use mosquito nets properly
            3) Help organize prevention activities
            
            Action = actively doing the prevention tasks
            
    Social:
        Community support = strongest overall factor
        If community supports, people:
                1) Work together (teachers, students, families)
                2) Encourage each other
                3) Join campaigns and follow rules

## 4: Model
### Social-Ecological Model (SEM)
A map of human behavior. It shaped by multiple layers.  
#### a ) Individual (YOU)
    Knowledge: factual understanding (Smoking causes cancer)
    Attitudes: positive/negative evaluation (I don't like smoking)
    Beliefs: what you think is true (may be wrong or culturally influenced)
    Self-efficacy: "Can I actually do this?" is the strongest predictor of behavior
    Risk perception: "How dangerous is this to me?"

#### b ) Interpersonal (RELATIONSHIP)
Your immediate social circle — family, friends, peers, partners.  
Humans are social creatures. Decisions are heavily influenced by what people around us think or do:

    Peer pressure: "If everyone around me does X, I should too."
    Social support: encouragement, advice, assistance
    Household power dynamics: A wife may need husbands approval to use contraception

#### c ) Community
Community = The people living around you or a social group
Broader environment — schools, neighborhoods, workplaces, religious institutions, local culture.  
Influence:

    Norms: What is socially acceptable?
    Leaders: Teachers, religious figures, local authorities
    Collective efficacy: Can the community work together to solve problems?

Example:

    In a village, if community leaders publicly promote handwashing → adoption skyrockets.
    If the community stigmatizes any method to prevent pregnancy use → knowledge alone fails.

    Communities reinforce or block behavior. Individual effort without community support is fragile.

#### Organizational (INSTITUTIONAL)
Formal organizations (inside a group or company) with rules and structures — schools, hospitals, government agencies, workplaces.  
Influence:

    Policies within institutions: rules, protocols
    Resource allocation: clinics stocked with medicine, schools with facilities
    Capacity: staff training, institutional commitment

Example:

    A school can offer nutrition programs → directly shapes children’s dietary habits.
    Hospitals providing free vaccination → enables individual action.

    Institutions enable large-scale behavior change by providing structure and resources.

#### Society (OUTER)
Government laws, national culture, economic policies, media, public norms.  
Influence:

    Legal frameworks: anti-smoking laws, mandatory vaccination
    Economic policies: subsidies for healthy foods
    Mass media: shapes cultural norms and awareness
    Deep-rooted cultural ideologies: gender norms, social roles

Example:

    Smoking ban in public places → changes social norms + reduces smoking rates
    National edutainment campaign → shifts attitudes across millions

    Policies shape the rules of the game — without supportive laws or culture, smaller interventions struggle.

### KABNP 
multi-layer decision system inside humans. 

#### a ) Knowledge
Objective information you have access to. 

Knowledge is necessary, but weak alone because humans don't act on facts, they act on what they believe + feel + what others expect. Reality you can know smoking kills but still smoke, So knowledge != action.  

#### b ) Beliefs
What you accept as "true", regardless of facts.  
Example:

    "Contraceptives cause infertility"

Beliefs equal to interpreted knowledge where beliefs come from:

    Culture
    Religion
    Family
    Experience
    Rumors / misinformation
    
Beliefs override knowledge. Example when you know vaccine works but believe it's dangerous, you reject vaccine.

#### c ) Attitudes
Your feeling toward something.  
Example:

    "I don't like contraceptives"

Formation:

    Attitude = f(Knowledge + Beliefs)
Example:

    Knowledge: "Contraceptives prevent pregnancy"
    Belief: "They cause harm"

    Attitude = Negative

Humans act emotionally, then justify logically, so attitude is closer to action than knowledge.

#### d ) Norms (SOCIAL FORCES)
What society expects you to do.  
Example:

    "Women shouldn't use contraceptives"
Types:

    Social norms → group behavior
    Gender norms → roles
    Cultural norms → traditions

Norms often dominate everything because humans fear rejection, shame & isolation more than risk, logic & facts.  
If (Attitude != Norm), norm usually wins,  
Example:
    
    You want to use contraceptives, but society disapproves, you don't do it

#### e ) Practices
Actual behavior repeated over time. Example using contraceptives.  
Practice != one-time behavior.
It is:

    habit + routine + social pattern

SBCC doesn't aim for intention, It aims for practice.  

#### f ) Flow Noted
```
        Norms
         ↓
Knowledge → Beliefs → Attitudes
         ↓
      Practice

System it's NOT linear, it's interactive + feedback loop.
```

Norms can override everything. Example : 

    Case 1 (Failure):
        Knowledge = correct ✔️
        Belief = correct ✔️
        Attitude = positive ✔️
        Norm = negative ❌
        Practice = NO

    Case 2 (Success even with wrong knowledge):
        Knowledge = wrong ❌
        Belief = wrong ❌
        Norm = strong ✔️
        Practice = YES
Example: eople know it's harmful, But norm supports it, practice continues

#### g ) Advanced 
You should think like this:
    
    Practice = f(K, B, A, N)
    Weight(N) > Weight(A) > Weight(B) > Weight(K)
Meaning:
    
    Norms = strongest
    Attitudes = second
    Beliefs = third
    Knowledge = weakest
Feedback Loop:

After behavior happens, Practice → reinforces Norms → shapes Beliefs → filters Knowledge.  
Example:

    Community practices something
    → becomes norm
    → becomes belief
    → becomes "truth"

### Power Of Norms (Most dangerous force)
Even if:

    Knowledge = correct
    Attitude = positive

If norm says NO → you won't do it
Real example:

    FGM (Female Genital Mutilation)

People know it's harmful, but still do it. This because social acceptance > logic.
Humans fear social rejection more than physical harm.

### Descriptive & Injuctive Norms
How brain processes :

    Should I do this behavior?:
        (Is everyone doing it?)  ← descriptive norm
        + (Will people judge me?) ← injunctive norm

Descriptive norm → tells you what is "normal"
Injunctive norm → tells you what is "acceptable"
        
#### a ) Descriptive 
Reality signal / What people actually do.  
Example:

    "Everyone litters"

Your brain interprets "This is common", "This is normal" & "I won't stand out if I do it". The result is you are more likely to litter, This is called Normalization effect
Even worse case when "Corruption is everywhere", Your brain:

    System already broken
    One more person doesn't matter
    I'd be stupid NOT to do it
    Don't be naive

The result will be corruption increases.
  
Deep truth, People don't want to be "the only one behaving differently" (humans are designed for survival through groups).

#### b ) Injunctive
"People disapprove of littering"
  
Your brain interprets:
    
    "If I do this, I get judged"
    "I might lose respect"
    "This is socially risky"

The result is you avoid the behavior. This is driven by fear of social punishment (rejection, shame, loss of status).

#### c ) Norms Conflict
This is where real behavior is decided.  

Situation:
    
    Descriptive: "Everyone litters"
    Injunctive: "Littering is wrong"
Which wins? Usually Descriptive norm wins, Because:

    Reality > words
    Observation > rules

"What people do is stronger than what people say".  

Theory that support this : Social Proof, Theory of Planned Behaviour, Social Cognitive Theory, Normative Social Influence & Cognitive Dissonance Theory.

#### d ) Norms Override Everything
Even if you know something is wrong, you believe it's bad, you want to do good, but if norm says opposite, you follow norm, Because humans are wired for social survival, group belonging & avoiding rejection. Deep biological truth is being rejected by the group is historically equal to death. So your brain prioritizes:

    Social safety > logical correctness

#### e ) Framework
When designing message, ask:

    What behavior do people think is common? (Descriptive)
    What behavior do people think is approved? (Injunctive)
    
Then:
    
    If bad behavior is common, Do not highlight it
    
Instead, shift attention to:

    positive minority
    social approval
    role models

You are not controlling behavior directly, You are controlling perception of reality because people don't act on truth, they act on what they think others are doing.

## Psychologies Theories
### 1 ) Theory Of Planned Behavior (TPB)
People don't act randomly, they act based on intention. And intention is built from 3 forces (attitude, social pressure & control). 

    Attitude + Social Pressure + Control → Intention → Behavior
TPB is about prediction, not guaranteed action.

#### a ) Attitude
Attitude is influenced by knowledge (facts) & beliefs (true or false)

So if beliefs are wrong then attitude becomes wrong.  
This can be:

    Knowledge → Beliefs → Attitude
    Beliefs → Attitude (directly)

Not just "like or don't like", It is your calculation of benefits vs costs.  
Example:

    Benefit: healthy body, better mood
    Cost: tired, time-consuming

If benefits > cost → positive attitude. Human tend to more do short term reward than long term reward (Temporal discounting).  

#### b ) Social Pressure
This is not just "others expect it"
It is what you think others expect + how much you care.  

Example:

    Your friends smoke
    Your family disapproves

Now conflict happens. You follow the group you value more (Social identity theory).  
Social pressure has 2 parts:

    Norm belief (What do others think?)  
    Motivation to comply (Do I care?)  

#### c ) Control
It's not about how hard something really is, It's about how hard you think it is.  
Example, eating healthy:

    Healthy food is available ✅
    You can afford it ✅

    Actual difficulty = EASY

But you think:

    "Healthy food is boring"
    "Too hard to change"

    You don't eat healthy
Its all in your mind the most hardest to control, you cant control external factor but you can control your mind.

    Control ≈ Self-efficacy + resources (time, money, access, tools, ooportunities)

#### d ) Intention 
This is the decision point, "I will do this".
But here is the trap, Intention != Behavior (always), Because unexpected barriers, environment changes & weak control. 

### 2 ) Social Cognitive Theory (SCT)
People don't just learn from experience, they learn by watching others. Behavior is a loop, not a straight line.  

#### a ) Observational Learning
You copy behavior from friends, influencers, parents & media.  
Example, You see someone getting rewarded for gym, you think "This works, I should try"

You don't copy everything, you copy when:

    Person is similar to you
    Person is successful
    Outcome is positive

#### b ) Self-Efficacy
"Can I actually do this?".  
This controls:

    Effort
    Persistence
    Resilience
Example two people:

    Same knowledge
    Same environment

    One succeeds, one fails

Why? Different self-efficacy.

#### c ) Environment Interaction
Environment is not passive.
It:

    Rewards behavior
    Punishes behavior
    Enables or blocks behavior


#### d ) Flow 
SCT core:

    Person => Behavior => Environment
##### i ) Person → Behavior
Your beliefs affect your actions

##### ii ) Behavior → Environment
Your actions change your surroundings  

Example:
You start exercising → others notice → gym culture grows

##### iii ) Environment → Person
Society shapes your thinking  

Example:
Everyone around you is lazy, you become lazy.

## 5: Sociology
Sociology in SBCC focuses on why people behave the way they do because of society, not just their own minds.
It explains external, structural, and relational forces that shape behavior.

### a ) Habitus
Your behavior, tastes, and decisions are so deeply shaped by your environment that they feel natural, even though they are learned.  

    Learned through: family, school, peers, culture, socio-economic class
    Operates: largely unconsciously
    Feels: "I'm just doing what comes naturally"

#### i ) Socialization
From birth, we absorb values, norms, and ways of acting.  
Example:  
children from wealthy families learn to eat gourmet foods, poor families may rely on cheaper, high-carb diets

#### ii ) Internalization
Over time, these behaviors feel automatic, not choices.
Example:  
saying "I'm just not a healthy eater" might really be "my habitus is shaped by my family and environment"

#### iii )
Reproduction
Habitus maintain social structures.  
What you "naturally do" reinforces class, culture, or community norms.  
The way you usually act (what feels "normal" to you) helps keep your group's habits and lifestyle the same.

### b ) Social Capital
Social capital is the power of your network, who you know, who trusts you, and how influence flows through relationships.  

#### i ) Trust
People are more likely to act on advice from someone they trust.  
Example, a health tip from a local respected leader is higher uptake than a random poster.

#### ii ) Relationships
The quality and density of your connections

    Close ties = strong influence
    weak ties = bridges to new ideas

#### iii ) Community Influence
People comply to norms in their network. Peer approval motivates behavior more than rational knowledge

Information travels faster along trusted networks than through general communication. Behavioral norms are enforced socially:

    if your friends start exercising, you're more likely to start
    Interventions leverage networks to "cascade" behavior change, Use people's connections so change spreads from one person to another, like a chain reaction.
