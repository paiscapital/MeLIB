# Theories in Public Relations


#### Main Question 

    How should organizations communicate with people, and what role should communication play in society?

Some believe PR exists to Control messages, Influence public opinion & Protect organizations. Others believe PR exists to Build relationships, Encourage dialogue & Create mutual understanding. Because PR has evolved over decades, many theories emerged to explain:

    How communication works
    How people react to messages
    How organizations interact with society

This lecture organizes PR theories along a continuum developed by Botan and Trowbridge. Think of it as a spectrum:

    Practitioner
         ↓
    Message
         ↓
    Organization
         ↓
    Public
         ↓
    Society

As we move rightward:

    Communication becomes less about control
    More about relationships
    More about participation
    More about societal impact

This is the intellectual evolution of PR.

## 1: Principles
### a ) Communication Creates Reality
Most people think:

    Reality → Communication

PR scholars often think:

    Communication → Perceived Reality

Example, Two companies experience the same crisis.

    Company A:
        "Everything is under control."

    Company B:
        "We made a mistake and are fixing it."

Same reality. Different communication. Different public reactions. Thus:
    
    Perception matters.

### b ) Organizations Depend on Publics
No organization survives alone. Organizations need:

    Customers
    Employees
    Governments
    Communities
    Investors

Without stakeholder support:

    No trust
        → No legitimacy
        → No survival

Therefore PR exists partly to maintain those relationships.

### c ) Power Exists in Communication
Communication is never neutral. Someone decides:

    What is said
    What is hidden
    Which issues matter
    Who gets heard

PR theories are often debates about Who should hold communication power?:

    The organization?
    The media?
    The public?
    Society?

This question drives nearly every theory in this lecture.

## 2: Theories
### a ) Role-Centred Theories
RCT believe that the success of public relations depends on the skills and actions of the PR practitioner. The practitioner's job is to create and spread messages that influence how the public thinks about an organization. Example :

    Journalist in Residence
    Hired Guns
    Court of Public Opinion
    Silver Bullet Theory

#### i. Journalist in Residence
Many early PR professionals were former journalists. Their strengths:

    Media contacts
    News writing
    Storytelling
PR success depended heavily on getting media coverage. Intuitive Analogy, Imagine knowing every teacher in school. If you need an announcement spread quickly:

    You already know who to call.

That is the advantage of former journalists. Modern PR professionals still benefit from:

    Media relations
    Storytelling
    Newsworthiness

#### ii. Hired Guns
PR practitioners sell communication services, the problem:

    Sometimes they prioritize clients over ethics.

This creates risks:

    Manipulation
    Propaganda
    Misleading information
Like a lawyer who defends anyone willing to pay. The skill itself is neutral, but the ethical use is the challenge. PR professionals constantly balance:

    Client Interest
        vs
    Public Interest

#### iii. Silver Bullet Theory
Assumes:

    One powerful message can instantly change public opinion.
This comes from early mass communication thinking. Like shooting a magic bullet into the audience's mind. Everyone reacts identically. Why It's Wrong is People are different. Different:

    Values
    Experiences
    Beliefs
Thus modern communication research rejects this simplistic assumption.

### b ) Channel-Centred Theories
Now attention shifts from:

    Who communicates
    to
    How communication works

These theories study:
    
    Messages
    Media
    Persuasion
    Information flow

#### Messages Are Not Reality

Most beginners assume:

    Reality = Message

Wrong. Channel-centred theories say:

    Reality
        ↓
    Message Construction
        ↓
    Audience Interpretation
        ↓
    Perceived Reality

The audience never sees reality directly. They only see the message, Example Company loses RM1 billion.

    Message A:
        "Company loses RM1 billion."

    Message B:
        "Company invests RM1 billion for future growth."

Same event, Different interpretation. This leads directly to Framing Theory Because whoever controls the frame often controls interpretation.

#### The Medium Changes the Message
The same message creates different effects depending on the channel. Example:

    Twitter/X
        ↓
    Fast
    Short
    Emotional

    Research Journal
        ↓
    Detailed
    Analytical
    Evidence-based

Same information with Different audience reaction. Imagine proposing marriage.

    Text message:
        "Will you marry me?"

    versus
    
    In person:

        "Will you marry me?"

Same words but Different impact. This is channel effect.

#### Attention Is Scarce
Modern humans are flooded with information. Every day people see:

    News
    Ads
    Social media
    Emails
    Videos

Thousands of messages. The problem is no longer:

    "How do I send information?"
The problem is:

    "How do I get attention?"
This is where:

    Agenda Setting
    Framing
    Rhetoric
    Persuasion
become powerful.

### Four Major Components
#### i. Message
The content itself. Questions:

    What is being said?
    What words are used?
    What information is included?
    What information is excluded?

Every message has:

    Denotation: Literal meaning.
        Example:
            "The company reduced 500 jobs."

    Connotation: Emotional meaning.
        Example:
            "The company streamlined operations."

Both mean layoffs. Different emotional impact. PR experts spend enormous time optimizing connotations.

#### ii. Media (Channel)
The pathway carrying the message. Examples:

    TV
    Newspapers
    TikTok
    Instagram
    Podcasts
    Press conferences
Every channel has built-in biases. A TikTok video encourages:

    Entertainment
    Simplicity
    Emotion

A scientific journal encourages:

    Evidence
    Logic
    Detail

Therefore:

Message Effectiveness = Message + Channel    
Not message alone.

#### iii. Information Flow
How information travels. Imagine:

    Organization
        ↓
    Journalist
        ↓
    Media
        ↓
    Public 
    
    or

    Organization
        ↓
    Social Media
        ↓
    Public
        
Different pathways create different outcomes. Why This Matters, Every step introduces:

    Filtering
    Interpretation
    Distortion

Information rarely reaches the public exactly as originally intended. Advanced Concept:

    Information Subsidies

Mentioned in your slides. PR practitioners create:

    Press releases
    Fact sheets
    Media kits

These are "pre-packaged information", They make journalists' jobs easier. As a result:

    PR influences news indirectly.

#### iv. Persuasion
The ultimate goal of many communication campaigns. Persuasion seeks to change:

    Knowledge
        ↓
    Attitude
        ↓
    Belief
        ↓
    Behavior
    
Example, Anti-smoking campaign.

    Step 1:
        Teach smoking is dangerous (Knowledge).

        ↓

    Step 2:
        Create concern (Attitude).

        ↓
    
    Step 3:
        Smoking is harmful (Belief).

        ↓

    Step 4:
        Quit smoking (Behavior).

#### Advanced Idea

The deepest idea behind Channel-Centred theories is: 
    
    Communication is not the transfer of information. Communication is the construction of meaning.
Most people think:

    Sender
        ↓
    Message
        ↓
    Receiver

Like sending a package. But advanced communication scholars think:

    Sender
        ↓
    Message
        ↓
    Interpretation
        ↓
    Meaning
    
Meaning exists inside the audience's mind. Not inside the message.

#### Press Agentry
Goal is to Get attention. Accuracy is often secondary. Success metric:

    Visibility

not necessarily truth. A circus promoter shouting:

    "Greatest show ever!"
The objective is attention. Modern examples:

    Clickbait headlines
    Celebrity publicity stunts

#### Media Relations
More balanced than Press Agentry. PR practitioners work cooperatively with journalists. Goal:

    Create mutually beneficial relationships.
Not a salesman. More like a long-term business partner.

#### Framing Theory
People do not react directly to facts. They react to how facts are presented. This is called a frame.  

A camera lens, The lens determines:

    What appears
    What stays outside the picture
Example, Unemployment:

    Frame A:
        "Thousands lost jobs."

    Frame B:
        "Economic restructuring creates future growth."

Same event, Different interpretation. PR professionals actively create frames.

#### Agenda Setting Theory
Media may not tell people what to think. But they tell people What to think about. Example, Teacher says:

    "This chapter will be on the exam."
Suddenly everyone pays attention. Importance is socially constructed. PR influences agendas through:

    Press releases
    Briefings
    Information subsidies

#### Rhetoric Theory
Studies persuasion through language. Words shape perception. Example, Politician says:

    "For the rakyat."
The word activates emotional associations. Language is software for the brain.

#### Persuasion Theory
Attempts to influence:
    
    Attitudes
    Beliefs
    Emotions
    Behaviors
Uses psychology heavily. Applications:
    
    Crisis communication
    Fundraising
    Political campaigns
    Community engagement

### c ) Organization-Centred Theories
This is where PR becomes strategic management, PR is no longer merely communication, PR becomes:

    A management function.
    Issues Management
Organizations continuously face emerging issues. PR scans the environment for:

    Complaints
    Trends
    Activist concerns
    Regulatory risks
before they become crises. Example, Weather forecasting:

    Detect storms before they arrive.

Early detection saves money and reputation.

#### Crisis Management
##### i) Situational Crisis Communication Theory (SCCT)
Developed by W. Timothy Coombs. Core Idea: 
    
    Different crises require different responses.
    Not every crisis should be handled the same way.
Example, Victim Crisis:

    Company not responsible.
    Response:
        Sympathy
        
Preventable Crisis:

    Company responsible.
    Response:
        Apology
        Compensation
Different diseases require different medicine.

##### ii) Symmetrical Theory
Developed by James E. Grunig and Todd Hunt.  
Asymmetrical Communication:

    Organization persuades public.
    Organization → Public
    
Symmetrical Communication:

    Organization and public learn from each other.
    Organization ↔ Public

Debate versus conversation.

##### iii) Excellence Theory
Excellent organizations create:

    Mutual understanding
    Trust
    Long-term relationships
PR contributes directly to organizational success. PR becomes a strategic advisor sitting at the executive table. Not merely a press-release writer.  
  
Crisis Management = Damage Control  
Actually, Issues Management:

    → Crisis Prevention
    → Crisis Response
    → Recovery
<br>

### d ) Public-Centred Theories
This is a major philosophical shift. Question changes from:

    "How can organizations influence people?"
    to
    "How do people experience organizations?"

#### i ) Co-Orientation Theory
Groups often misunderstand each other. The theory studies:

    What I think
    What you think
    What I think you think
Example, Two friends arguing through assumptions:

    Most conflict comes from perception gaps.

#### ii ) Dialogic Theory
Inspired by Jurgen Habermas. Communication should be genuine dialogue, Every stakeholder should have equal opportunity to participate. Best arguments win, not strongest power. Example, Round-table discussion is Every voice matters.

#### iii ) Situational Theory of Publics
Developed by James E. Grunig.  
Four Public Types:

    Non-Public
    Latent Public
    Aware Public
    Active Public
Fire alarm example:

    Non-Public (No fire): The issue does not affect them, and they don't care.
        
    Latent Public (Smoke is hidden): The problem affects them, but they don't know it yet.

    Aware: Smoke recognized.

    Active: People running toward exits.

Different publics require different communication strategies.

#### iv ) Relationship Management Theory
Relationships are assets, Trust must be maintained continuously. Example, Friendship:

    You cannot build trust once and ignore it forever.

#### vi ) Co-Creational Theory
Meaning is not created by organizations alone. Meaning is jointly created.

    Organization + Public
Example, Writing a book together rather than reading a finished book.

<br>

### e ) Organizational Roles in Society
This is the most advanced level, The unit of analysis becomes society itself. PR asks:

    How can organizations contribute to communities and social systems?

#### i ) Community Building Theory
Organizations should not be communication centers. They should become:
    
    Contributors to community well-being.
Examples:

    Long-term CSR
    Community partnerships
    Social development

Example, A tree in a forest. A tree survives only if the ecosystem survives. Modern PR success is increasingly measured through:

    Trust
    Social legitimacy
    Community impact
not merely profit.
