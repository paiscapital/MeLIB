# Media Selection & Media mix


At surface:  
&emsp;    Media selection = the process of choosing the most effective platforms to deliver advertising messages to the target audience.    
&emsp;    Media mix = combine platforms

Advanced:  
&emsp;    Optimizing information delivery under constraints (budget, attention, psychology, geography).     
&emsp;    "Where is the highest probability that my message changes behavior per RM spent?"
<br>
#### Nature of Advertising
    Media Selection - Choosing the most suitable platform (e.g., TikTok, TV).  
    Media Objectices - What the brand wants to achieve (awareness, sales, engagement).  
    Media Strategy - How and where to deliver the message.  
    Media Planning Process - Step-by-step process of selecting and scheduling .  
    Media Choices Benefits - The advantages gained from using specific media.  


## 1: Media Selection 5 Variables

##### Genz != TikTok
Gen Z tends to prefer short content & like fast, visual, high-stimulation formats. But Those behaviors exist across multiple platforms not just tiktok. Short-form content exists in TikTok, Instagram Reels & YouTube Shorts. So if you say:

    "Gen Z → TikTok"

You are arbitrarily limiting your reach. So platform choice = behavioral match, not demographic. Platform determines distribution environment & Behavior determines content style.  


##### Scenario A
    Gen Z → TikTok → post content there

Problem:

    You ignore Instagram Reels (huge overlap)
    You ignore YouTube Shorts (growing fast)
    You ignore timing/context differences
Result: under-optimized reach


##### Scenario B
    Gen Z behavior = short, fast, visual → create that content → distribute across multiple platforms

Result:

    Higher reach
    Better efficiency
    Lower risk

### b ) Objective (What you want to change)
| Objective  | Mental State Transition             | Media       |
| ---------- | ----------------------------------- |-------------|
| Awareness  | "I don't know" → "I recognize this" | TV, TikTok  |
| Engagement | "I see it" → "I care about it"      | IG, TikTok  |
| Conversion | "I care" → "I act now"              | Shopee, Ads |

#### i ) Awareness = Recognition
This is the lowest level of cognition, where the goal is not persuasion but merely to trigger recognition and a sense of familiarity in the mind. The brain works:

    1) Fast processing (System 1 Thinking)
    2) No Deep Thinking
    3) Visual Repitition Builds Memory

TikTok platforms works because its high exposure, repitition & passive consumption. The brain also stores Logo, color & sound but not the prices, features & decision.

    Awareness != persuasion
    It is just memory encoding

#### ii ) Engagement = Interaction
At this stage, you don't just notice something & you start to feel it or react to it. The brain process:

    1) Dopamine (Likes, comments, shares)
    2) Emotional trigger (Funny, relatable, aesthetic)
    3) Light Thinking

Instagram and TikTok works because its interactive (comment, share), Social validation (likes) & Algorithm amplifies engagement. This change the brain from passive viewer to active participant.

    Engagement builds connection, not decision

#### iii ) Conversion = Action
This is the hardest stage because the goal is to make people act immediately, such as buying something, clicking a link, or spending money. The brain process:

    1) Rational + emotional combined
    2) Risk Evaluation
    3) Urgency trigger

Shopee and Ads work because the user already has intent, environment supports action (buy button) & resistence is low.

    "Click this link and then press buy button"

### c ) Message Nature (How complex it is)
How difficult it is for the audience to understand it, believe it, remember it & act on it. Its about cognitive load:

    How much mental effort the brain needs to process the message.

#### Complexity to changes media choices
Different media have different attention capacity.
| Media        | Attention Depth | Brain Load Capacity          |
| ------------ | --------------- | ---------------------------- |
| Billboard    | very low        | instant recognition          |
| TikTok       | low-medium      | fast emotional reaction      |
| TV / YouTube | medium-high     | storytelling possible        |
| Seminar      | very high       | explanation + trust building |

The more cognitive work required, the more time and attention you need from the audience. 

| Message Type                       | What the brain needs          | Best Media Type                               |
| ---------------------------------- | ----------------------------- | --------------------------------------------- |
| Simple (discount, promo)           | Instant recognition           | Billboard, TikTok, banners                    |
| Emotional (branding, lifestyle)    | Feeling + association         | TV, YouTube, Instagram video                  |
| Complex (insurance, finance, tech) | Understanding + trust + logic | Long-form video, articles, webinars, seminars |

### d ) Geography (Where decision happen)
Basic:

    Local vs international
    Local = nearby audience
    International = global audience

Advanced:

    Geography = where decision probability becomes highest

Not where people see your ad. But where they are most likely to notice, trust, act & purchase. 
                                                                                                                                                                     
    decision environment targeting

#### Decide happen when 3 things align
##### i ) Physical proximity
    They are near the action point (cafe, mall, stores)

##### ii  Mental readiness
    They are already thinking about the need (Hungry, thirsty, bored)

##### iii ) Low friction environment (a situation where it is very easy for someone to take action)
    It is easy to act immediately (no long delay, no travel barrier)

#### Example
Scenario A: Billboard near café

    Person driving nearby
    Sees café ad
    Feels hunger/thirst
    Café is 200m away
    Result: High conversion probability

    Because:

        Distance = low
        Effort = low
        Action = immediate
This is high decision proximity.  
  
$
P(conversion) = f(proximity, intent, friction)
$

proximity = how near they are to action location  
intent = how much they want it right now  
friction = effort needed to act

### e ) Cost Efficiency (How much impact per RM)
CPM = Cost Per Mille (1000 impressions)

    "How much do I pay to show my ad to 1,000 people?"

Example:

    RM100 CPM → you pay RM100 for 1,000 views
CPM only measures:

    Exposure (views)
    Not what happens after exposure

So CPM answers:
    
    "Did people see it?"

But Not:

    "Did people care or act?"

#### CPM Ignores 3 Critical Layers

##### a) Attention Quality
Not all views are equal
| Platform           | Attention Type   |
| ------------------ | ---------------- |
| TikTok             | fast, distracted |
| YouTube long video | focused          |
| Google Search      | intentional      |

1,000 TikTok views != 1,000 Google search views.  
CPM treats them as equal, But in reality they are not.  

##### b ) Conversion Probability
CPM doesn't ask:

    "What is the probability that a viewer buys / clicks / acts?"

Example:

    TikTok viewer:
        0.1% chance of buying
    
    Google Ads viewer:
        5–20% chance of buying
Even if TikTok is cheaper per CPM, it may produce less profit.  

##### c ) Intent Level
How ready someone is to act.
| Platform      | Intent          |
| ------------- | --------------- |
| TikTok        | Low (scrolling) |
| Instagram     | Medium          |
| Google Search | High            |

##### d ) Lowest CPM != Best
If RM10 CPM is cheaper than RM50 CPM, use RM10.  
That is WRONG because:

    You are optimizing COST, not OUTCOME.
Example:
    
    TikTok Ads:
        CPM = RM10 (cheap)
        100,000 views = RM1,000
        Conversion rate = 0.1%
        → 100 sales

    Google Ads:
        CPM = RM50 (expensive)
        100,000 views = RM5,000
        Conversion rate = 5%
        → 5,000 sales

| Platform | Cost | Sales | Efficiency |
| -------- | ---- | ----- | ---------- |
| TikTok   | Low  | Low   | Weak       |
| Google   | High | High  | Strong     |

## 2: Mix Media 
Basic: 
    
    Combination of different channels

Advanced: 

    Media mix = multi-layer funnel system

    Layer 1 → Awareness
    Layer 2 → Interest
    Layer 3 → Conversion
    Layer 4 → Retention

Example:

| Stage      | Media            | Function         |
| ---------- | ---------------- | ---------------- |
| Awareness  | TV               | Mass reach       |
| Interest   | TikTok           | Engagement       |
| Conversion | App notification | Immediate action |
This is not random. This is sequential psychological triggering.

## 3: Characteristics of media
Different types of advertising media have different characteristics, advantages, and limitations.

Each media has 3 hidden properties.  

#### i ) Attention Depth

| Media   | Depth  |
| ------- | ------ |
| TikTok  | Low    |
| TV      | Medium |
| Article | High   |

#### ii ) Interrupt vs Intent
| Media         | Type           |
| ------------- | -------------- |
| Billboard     | Interrupt      |
| TikTok        | Semi-interrupt |
| Google Search | Intent         |
Intent media = higher conversion

#### iii ) Outdoor / Mural Advertising
Billboards, wall murals.  
Advantages:
    
    Eye-catching
    High visibility
    Good for awareness
    E.g : Billboard di jalan raya Kota Kinabalu & Coca-Cola wall ads

#### iv ) Press Advertising
Newspapers, magazines.  
Advantages:
    
    Wide reach
    Easy to distribute
    Quick results
    E.g : Berita Harian, The Star & The New York Times

#### v ) Radio Advertising
Advantages:
    
    Affordable
    Can target specific audience
    Good for reminders
    E.g : ERA FM, Hitz FM & Spotify Ads

#### vi ) TV Advertising
Commercials, sponsorships.  
Advantages:

    Large audience
    Strong visual + audio impact
    E.g : TV3, Astro

#### vii ) Online Advertising
Social media, websites, apps.  
Advantages:
    
    Highly targeted
    Interactive
    Real-time feedback
    E.g : TikTok Shop, Shopee Live, Google Ads & Meta Ads

#### viii ) Memory Retention
| Media | Retention |
| ----- | --------- |
| Video | High      |
| Audio | Medium    |
| Text  | Low       |
