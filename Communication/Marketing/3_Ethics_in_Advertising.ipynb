# Ethics in Advertising

Ethics : rules about what is right and wrong that guide how people behave. They include being honest, fair, responsible, and respectful toward others.  
Ethics = honesty, responsibility, respect, trust building.  

    Long-term system stability mechanism

Because:
    
    deception → short-term gain, long-term collapse
    trust → slow gain, long-term survival

   
Advertising ethics = Constraint layer on persuasion systems. You are trying to Influence behavior (buy, click, believe), But under constraints (law, society, psychology).  


## 1: Truth in Advertising
Deception = making false or misleading statements in advertising
Puffery = non-verifiable, subjective exaggeration designed to influence emotion without making a legally testable claim. (vague) "The best coffee in the world"  

## 2: Issues in Advertising
### a ) Advertising to Children
Advertising to children is controversial because children are not fully capable of understanding persuasion.

    Children are easier to influence → society tries to protect them

1. Pro-regulation side:

        Children are vulnerable  
        Ads manipulate them unfairly  
        Needs protection  
   
2. Anti-regulation side:

        Advertising = free market communication
        Parents should control exposure
        Businesses need freedom

### b ) Advertising Controversial Products
Products that create social harm risk.  
Examples:

    Tobacco (health damage)
    Alcohol (addiction / social harm)
    Gambling (financial addiction)
    Pharmaceuticals (medical influence risk)

So the core issue is:

    These products can legally exist, but their promotion is ethically dangerous.

Tabacco advertising bans. The Goal is to reduce:

    exposure
    normalization
    behavioral adoption
Mechanism:
    
    Remove mass media visibility → reduce social acceptance → reduce consumption

##### But they adapt using Indirect Advertising.
Promoting brand through other products instead of the harmful product itself.  
Example, cigarette brands promote:

    travel
    clothing
    restaurants

This is not "breaking the law". It is:

    Regulatory bypass via identity transfer

How it works is Instead of:

    "Buy cigarettes"

They do:

    "Buy lifestyle identity associated with cigarette brand"
They don't sell the product. They sell status, identity, lifestyle. This is extremely important in modern marketing systems.

#### 3 Competing Forces
1. Industry Forces

       Maximize Influence: Expand reach, Bypass restrictions, use indirect branding 
2. Government Forces

        Limit Harm: Bans, Regulation, Advertising laws
3. Consumer Forces:

        Vulnerability Layer: Children, Addict, Low information users

## 3: Stereotyping in Advertising
Stereotypes = general beliefs about traits, roles, behavior patterns of groups.  

    Women → emotional, caring, domestic
    Men → strong, logical, leader
    Elderly → weak, dependent

These are not facts about individuals, They are compressed social assumptions.

    Cognitive shortcuts for social prediction

Humans do this because:

    Brain cannot analyze every person deeply
    So it builds "default models" of groups

#### a ) Women in Advertising
Women often shown as beauty-focused, domestic, nurturing. Men shown as strong, independent, achievement-oriented.

    Role assignment in narrative systems

| Gender | Narrative Role |
| ------ | -------------- |
| Women  | Care + emotion |
| Men    | Power + action |

Example:

    Women -> Nursing (role) -> Family chef (Identity Framing) -> Product (Emotion)

    Create delicious meals your family will love, make every moment at the table special with this sauce.

#### b ) Senior Citizens (Age)
Older people often shown as weak, dependent, losing memory.  

    Age-based capability compression
    Society compresses elderly into "declining system model"
Real cognitive bias, Humans associate:

    age ↑ → capability ↓
    Even when it is not always true.

Advertising effect, Elderly are rarely shown as:

    active
    powerful
    independent

Instead:

    dependent
    nostalgic
    fragile

Stereotype (Older people are less productive) → belief (Age means decline in ability) → social treatment (Older workers are overlooked for promotions) → self-identity (I'm too old for this tech) → reinforcement (Older people are excluded from tech roles)  

#### c ) Racial & Ethnic
Minorities often reduced to jokes or background roles.  
Examples:

    "Angry Black Girl"
    "Black Man in Trainers"

    Identity reduction in media systems
Media turns real people into "short descriptions" instead of showing everything about them.  
Example:

    A student who is hardworking, funny, and good at sports gets shown in media only as "the nerd" or "the athlete".

this is dangerous because it can make people see others in a limited and unfair way, leading to three problems such as Expectation bias (expect before knowing the truth),  Social prejudice (Unfairly think/judging) & Unequal representation (show only young dr not older dr)  

#### d ) Simple Summary
Stereotypes = mental shortcuts about groups.  
Advertising uses them to communicate faster.  
They increase persuasion efficiency.  
But they reduce individual representation.  
And can reinforce social bias loops.  

## 4: Advertising Code In Malaysia
#### a ) Advertising Standards Advisory (ASA)
ASA established in 1977 & Provides independent scrutiny of self-regulation system.  
ASA is like a system where advertisers "police themselves", but an outside group watches to make sure they don’t break the rules.  

There are 3 actors:

    Advertisers (companies)
    ASA (self-regulation body)
    Government (legal fallback)

    Industry → self-polices → ASA reviews → government intervenes only if needed

This exists because pure government control is:

    slow
    expensive
    not scalable for thousands of ads

So ASA = Decentralized (not truly) compliance filter

#### b ) Cultural & National Identity Requirements
The rules must:

    Reflect Malaysian culture
    Reflect multi-racial society
    Support Rukun Negara philosophy 

    Narrative alignment with national ideology

Rukun Negara = ideological constraint system.  
It ensures ads do not:

    break racial harmony
    distort social unity narrative
    destabilize identity perception

System model:
    
    Advertisement message → must pass cultural filter → must not break national cohesion narrative

This is Social coherence optimization constraint. Because ads shape perception of:

    race
    identity
    belonging
    hierarchy


So Malaysia forces:

    "No fragmentation of identity narratives"
    Show the whole person, not just one part.

#### c ) Anti-Streotyping Rules
Should not assume someone's job or role based on their race or gender.   
This rule exists because advertising naturally tends to:

    compress groups into simple behavioral models

But that creates:

    bias reinforcement
    identity distortion
    social segmentation

System model, Without rules:

    Group (Women) → simplified stereotype (Women = Emotional) → repeated exposure (Ads, Movies, Socmed) → normalized bias (This is true)

With rules:
    
    Group (Women) → diverse representation enforced (Media show many different roles) → bias disruption (Women can do many different things)

This is NOT just ethics. It is bias control in mass communication systems. Very similar to:

    fairness constraints in AI models
    anti-overfitting rules in machine learning



#### d ) Core Principles of Malaysian Advertising Ethics
1. Responsibility to society : Ads must not harm society 
2. Transparency : Ads must be clearly identifiable 
3. Fair competition : No misleading market advantage 
4. Trust protection : Ads must not damage confidence in advertising system 

This creates 4 stability constraints:

| Principle        | System role         |
| ---------------- | ------------------- |
| Responsibility   | social safety       |
| Transparency     | information clarity |
| Fair competition | market stability    |
| Trust protection | system legitimacy   |

Advertising is not just about selling products, it helps build trust between people and brands.

#### e ) Law Layer
1. Communications & Multimedia Act 1998
2. Trade Description Act 1972
3. Contracts Act 1950
4. Electronic Commerce Act 2006
5. Medicine Advertising Act 1956
6. Penal Code
7. Indecent Advertisement Act 1953
8. Food Act 1983 

This is a stacked enforcement system:

    Layer 1: Soft control = ethics & ASA
    Layer 2: Legal control = advertising laws
    Layer 3: Criminal control = Penal Code (serious violations)



System model:

    Ethics (soft)↓ASA (semi-hard)↓Law (hard)↓Penal Code (punitive)
This creates graduated enforcement pressure, meaning:
    
    most violations handled socially
    extreme cases handled legally

#### f ) Final Model

Advertising System:

    Ethical Rules (behavior) + Cultural Constraints (identity) + Legal Enforcement (law) + Regulatory Bodies (ASA) + Market Principles (competition)
