# Protection Motivation Theory

A psychological theory developed by R.W. Rogers (1975, 1983) to explain why people change risky behaviour or protect themselves from harm (Mental cost-benefits analysis).

    smoking cessation
    vaccine uptake
    cybersecurity behaviour
    health campaigns (HIV, obesity)

People decide to protect themselves based on 2 cognitive appraisals:

## 1: Threat Appraisal (How dangerous is it?)
This includes:

    Perceived Severity → "How bad is the outcome?"
    Perceived Susceptibility → "How likely will it happen to me?"
    Rewards of unhealthy behaviour → "What do I gain from doing the risky behaviour?"
    self-efficacy: How confident a person is that they can perform the protective action.

Example (Smoking):

    Severity: lung cancer is deadly
    Susceptibility: "I might get it because I smoke daily"
    Reward: stress relief / social acceptance

#### Transforming PMT into a weighted utility model
Severity (S) = 0 to 1 (0.8)  
Probability (P) = 0 to 1 (0.7)  
Risk value = S × P (0.8 x 0.7)  
Short-term reward (R) = 0 to 1 (0.7)  
Self-efficacy (E) = 0 to 1 (0.5)  
α (alpha) = short-term bias multiplier
  
Formula:  
##### Risk Pressure (avoid behaviour)
$$
Risk = S * P
$$
##### Reward Pressure (continue behaviour)
$$
Reward = R * α
$$
##### Decision Index
$$
Decision = Reward_{effective} - Risk
$$

### Code

```python
# -----------------------------
# Protection Motivation Theory (PMT) Model
# -----------------------------

S = 0.8   # Severity
P = 0.7   # Probability
R = 0.25   # Short-term reward
E = 0.5   # Self-efficacy (not used in basic equation yet)
alpha = 2.0  # short-term bias multiplier

# -----------------------------
# Risk Pressure (Avoid Behaviour)
# -----------------------------
Risk = S * P

# -----------------------------
# Reward Pressure (Continue Behaviour)
# -----------------------------
Reward_effective = R * alpha

# -----------------------------
# Decision Index
# -----------------------------
Decision = Reward_effective - Risk
Coping = E

# Adjusted decision (more realistic PMT version)
Decision_PMT = (Reward_effective * (1 - Coping)) - Risk


print("Severity (S):", S)
print("Probability (P):", P)
print("Enhanced Decision:", Decision_PMT)
print("Risk (S × P):", Risk)
print("Reward (R × α):", Reward_effective)
print("Decision Index:", Decision)

if Decision > 0:
    print("Behaviour: IGNORE RISK (continue behaviour)")
else:
    print("Behaviour: AVOID RISK (change behaviour)")
```

## 2: Coping Appraisal (Can I deal with it?)
This includes:

    Response Efficacy → "Does quitting actually help?"
    Self-Efficacy → "Can I successfully quit?"
    Response Cost → "How hard/expensive is it to quit?"
Example:

    Quitting reduces disease risk (efficacy)
    "I believe I can quit with support" (self-efficacy)
    Withdrawal symptoms make quitting hard (cost)
<br>

## 3: Final Decision (Protection Motivation)
After evaluating both, if threat is high + coping ability is high, person changes behaviour. But if:

    high threat but low coping → person avoids action
    low threat → no motivation

## 4: Real-World Example (Smoking)
The study aimed to test whether Protection Motivation Theory (PMT) can explain and predict smokers intention to quit smoking among adults in China.  

Method:

    Design: Cross-sectional survey
    Sample: Adults smokers from 26 provinces in China
    Data collection: Face-to-face questionnaires (Qualitative)
    Analysis method: Structural Equation Modelling (SEM)

Key Findings:
##### 1. PMT successfully predicts quitting intention
    Smokers with stronger PMT factors were more likely to intend to quit.
##### 2. Coping appraisal is stronger than threat appraisal
    Believing "I can quit successfully" and "quitting works" had a stronger effect than fear of disease.

Major psychological drivers:
    
    Self-efficacy (confidence to quit)
    Response efficacy (belief quitting helps health)
    Perceived severity & vulnerability (fear of harm)
    Response cost and rewards of smoking reduce quitting intention

Model explanation power:

    PMT explained about 12% of variation in quitting intention

PMT is a valid framework for explaining smoking quitting intention, but coping mechanisms (confidence + ability to quit) are more important than fear-based motivation alone.  

People don't quit smoking just because they fear disease, they quit when they believe they can successfully quit and it will actually work.
