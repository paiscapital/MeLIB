# Social Behavior Change Communication (SBCC)

Using communication to change people's behavior (for better life/health). Not just telling people information, but helping them understand, feel motivated and actually change what they do.  
Example, instead of just saying:

    "Don't smoke"
SBCC tries to:

    explain why people smoke
    change social pressure
    support them to quit
    make environment support quitting

Why people don't change even if people know something is bad an they still do it, Why? Because behavior depends on:

    psychology (habits, emotions)
    society (friends, family)
    environment (money, access)
    culture (what society accepts)
    rules/laws (government)

So information alone is not enough.  
SBCC mix many fields like Psychology (how people think), sociology (how society affects behavior), communication (how messages are delivered) & public health / development (real-world problems)
<br>

## Evolution 
<br>

### 1: IEC (1970s)
Information, Education, Communication (IEC). Basic ideas is "If people know the right thing → they will do the right thing", so IEC do posters, ads, campaigns & school lessons.  
Example:

    "Wash your hands to prevent disease"
    
    You know studying is important
    But you still procrastinate (intentionally delaying or postponing what need to be done)

Reality:

    People already know
    But they don't act

  
#### KAP Gap
Knowledge → Attitude → Practice (KAP)
IEC assumes:

    Knowledge → Attitude → Behavior
But in reality:

    Knowledge ≠ Behavior
This is the KAP Gap and can have cognitive dissonance.
<br><br>

#### IEC Fails
a ) Habit > Knowledge  
Behavior is automatic  
Example:

    You know junk food is bad
    But habit wins
Brain uses automatic system, not logic.  
  
b ) Social Pressure  
Humans follow people, not facts (Social proof, people follow others to decide what is "correct").  
Example:

    Everyone around you doesn't wash hands
    You follow them
  
c ) Lack of Ability / Resources  
Sometimes people literally can't act  
Example:

    No clean water → can't wash hands
    No time → can't cook healthy

d ) Low Perceived Risk (its not dangerous for them)  
People think:

    "It won't happen to me"
They don't think the risk is real for them.  

c ) No Immediate Reward  
Brain prefers short-term reward & not long-term benefit.  
Example:

    Smoking feels good now
    Cancer is later → ignored


#### Model
$$
\begin{aligned}
B &= f(I, C, E, S, P) \\[6pt]
I &= \text{Information (knowledge level, 0–1)} \\
C &= \text{Incentives (reward/punishment strength)} \\
E &= \text{Environment (ease/difficulty)} \\
S &= \text{Social norms (peer pressure strength)} \\
P &= \text{Psychology (motivation, habit strength)}
\end{aligned}
$$
#### Code

```python
import math

I = 0.8  # (they know a lot)
C = 0.2  # (low incentive)
E = 0.3  # (hard environment)
S = 0.6  # (moderate social pressure)
M = 0.4  # (low motivation)

w1 = 0.5 # Information matters, but only half as strong as a "full unit"
w2 = 1.0 # Incentives are stronger, full impact
w3 = 0.8 # Environment has medium-strong impact
w4 = 1.2 # Social norms are the most important factor
w5 = 1.0 # Psychology/motivation is also important

def P(z):
    return int((1 / (1 + math.exp(-z))) * 1000) / 1000

z_score = (I*w1) + (C*w2) + (E*w3) + (S*w4) + (M*w5)\
print(P(z_score))
```

### BBC (1990s)
Changing individual behavior by influencing their mind, not just giving them facts.  
Instead of saying "Here's what to do" → BCC asks "Why would someone choose to do it or not? What internal factors prevent action?"  
<br>

#### Behavioral Psychology Model
a ) Attitude  
Do they believe this action is good or bad?

    Positive attitude → more likely to adopt behavior
    Negative attitude → resistance, even if they know facts

b ) Fear / Perceived Risk  
How real do they see the danger?
Health Belief Model uses this: 

    perceived severity + perceived susceptibility → motivates action

c ) Self-Efficacy / Confidence  
Do they feel capable of performing the behavior?

    Low confidence → paralysis, procrastination, avoidance
  

Handwashing example:
IEC says "Wash your hands" → only info, no psychological targeting.
  
BCC says "Wash your hands because it prevents your child from getting sick"
Targets emotion and motivation → parental concern + perceived risk
<br><br>

### SBCC (2000s)
Function of Environment
Think of behavior B as a function:
$$
B = f(I,F,C,S)
$$ 
Where:  

    I = Individual factors (knowledge, beliefs, skills, self-efficacy)
    F = Family / interpersonal influences (peer pressure, norms at home)
    C = Community / social norms (what society accepts, local culture)
    S = System / structural factors (laws, policies, accessibility, cost)
  
Behavior is emergent, changing one variable alone rarely works because all levels interact. Traditional approaches target only I (individual knowledge), so impact is low. SBCC targets all levels simultaneously, creating a synergistic effect. This is like multi-factor optimization, if any level blocks change, the outcome is limited. Each level reinforces the others → higher probability of sustained behavior change.  
  
SBCC is dynamic, not static. Changing one level often feeds back into others:

    Family support (F) increases self-efficacy/confidence (I)
    Smoke-free policies (S) shift community norms (C)
    Media campaigns (C) make quitting socially desirable → reduces peer pressure (F)

Think in systems theory terms, it's networked behavior intervention.


SBCC as Behavior "Engineering":

    SBCC is essentially strategic ecosystem design.
    Instead of blaming individuals, we restructure the environment:
        Reduce friction for positive behavior
        Increase friction for negative behavior
    It's a control system: feedback + reinforcement loops guide behavior.

#### Code

```python
# Focus : Tanzania Stunting Prevention Communication Program
# Reduce child stunting (malnutrition)
# Outcome Var : Y = Child nutrition status (stunted = 1, normal = 0)
# Data Needed : 
"""
data = {
    'Individual (I)': ['Mother knowledge score (0–100)', 'Breastfeeding practices', 'Diet diversity'],
    'Family (F)' : ['Father involvement (binary:1-0 / scale:range of value 1-6 or score)', 'Decision-maker in household'],
    'Community (C)' : ['Social norms (survey: % agreeing with good feeding)', 'Media exposure (radio/TV frequency)'],
    'System (S)' : ['Access to health clinics', 'Availability of clean water', 'Government programs']
}
"""
# Multi-Factor dataset : X = (I,F,C,S)
# Model : Logistic Regression
# P(Y=1) = 1 / (1 + e^(-(β0 + β1·I + β2·F + β3·C + β4·S)))
# p = 1 / (1 + math.exp(-(beta0 + beta1*I + beta2*F + beta3*C + beta4*S)))
# Euler growth/decay unlimited : 
#    Growth → slow → fast → huge (interest finance)
#    Decay → fast → slow → tiny (Ice melting)
# Beta (in regression) : Coefficient or Weight (how strongly a variable affects the outcome / the effect size of that factor)
import math

# Stunting case
I = 5 # 0-10
F = 1 # 0/1 yes/no
C = 7 # 0-10
S = 1 # 0/1 yes/no access to clinic

# Beta Stunting
b0 = -2
b1 = -0.3
b2 = -0.5
b3 = -0.2
b4 = -0.8
stun = round(1 / (1 + math.exp(-(b0 + b1*I + b2*F + b3*C + b4*S))), 3)

# Healthy case
#Beta for healthy
h0 = 2
h1 = 0.3
h2 = 0.5
h3 = 0.2
h4 = 0.8

healthy = round(1 / (1 + math.exp(-(h0 + h1*I + h2*F + h3*C + h4*S))), 3)

Y = healthy - stun
print(f"""
Stunting val : {stun}
Healthy val : {healthy}
Y val : {Y}
""")
```
