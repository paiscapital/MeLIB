# Introduction to Advertising

#### Traditional:
Paid, non-personal communication about a product/service by an identified sponsor.  
    
#### New
Paid communication designed to persuade action now or later

A paid, planned communication used by a company or organization to inform, persuade, or remind people about a product, service, or idea.  

Advertising = "Pay money to send a message to influence people" (Capital → Media → Influence → Behavior Change)  

## 1: Objective
### a ) Inform Objective
Giving new, clear, and useful information to the audience so they understand a product, service, or idea. This is usually used when a product is new, a market is not familiar & there is complex information (Hard to understand, Technical & New or unfamiliar).  
#### i ) When Informative Advertising is Important
Stage refer to introduction Stage (Product Life Cycle). In marketing, products go through stages:

    Introduction → Growth → Maturity → Decline

Informing is most important at the beginning (Introduction), Because people:

    Don't know the product exists
    Don't understand how it works
    Don't trust it yet

##### Reducing Uncertainty
Informative ads are powerful because they reduce uncertainty.

In consumer psychology, People avoid buying when they feel confused, unsure & risky.
So advertising must increase knowledge, build understanding & reduce perceived risk.

| Inform               | Persuade                           |
| -------------------- | ---------------------------------- |
| Focus on facts       | Focus on emotions                  |
| Educate audience     | Convince audience                  |
| "This is what it is" | "This is why you should choose it" |

#### ii ) Types of Informative Advertising
1. Pioneering Advertising

        Introduces a completely new product category
    Example, first electric cars (people need explanation: What is it? How to charge?)

2. Competitive Informative Advertising

        Explains differences between brands
    Example, "This phone has better battery than others"
<br>


### b ) Persuasive Advertising
Persuasive advertising is about changing people's attitudes and behavior, not just informing them.

    "Make people prefer your brand over all others"

#### i ) Emotional
People often buy based on feelings, not logic.

    Happiness: "This product makes you feel good"
    Fear: "Without this, you might have a problem"
    Love/Belonging: "Be part of this group"

Example: Shampoo ad shows confident, beautiful person, you feel you can be like them.

#### ii ) Rational
Uses facts and logic:

    Better quality
    Lower price
    More benefits

Example: "Our shampoo has 3x more vitamins than others"  

#### iii ) Unique Selling Proposition (USP)
What makes your product different and better

Example: "Only shampoo with natural herbal formula"  

#### iv ) Social Proof
What makes your product different and better

Example: "Only shampoo with natural herbal formula"

#### v ) Repetition

The more you see it, the more you believe it.  

#### vi ) Persuasion Strategies
1. Comparative Advertising

        Compare directly with competitors
    "Better than Brand X"

2. Scarcity Principle

        Create urgency
    "Limited time offer!"

3. Authority Principle

        Use experts
    "Recommended by dermatologists"
<br> 

### c ) Reminder Advertising
Reminder advertising is used when people already know the product, but you want them to:

    Not forget it
    Keep buying it
    Stay loyal
Main goal is "Stay in the customer's mind all the time"

1. Memory Reinforcement

       Repeated exposure strengthens memory
    The more you see a brand, the more familiar it feels

2. Brand Recall vs Brand Recognition
    Brand Recall: You remember it without help

            "Name a soft drink", you say Coca-Cola
    Brand Recognition: You recognize when you see it

            You see the logo, "Oh, Coca-Cola!"

   
4. Top-of-Mind Awareness (TOMA)
    The brand that comes to your mind first

            This is the ultimate goal of reminder ads


Big Brands Use Reminder Ads, Even famous brands like Coca-Cola still advertise because:

    Competition is strong
    Customers can switch brands easily
    Visibility = power
#### Type of Advertising
1) Brand Advertising
        Long-term identity building
        Example: TV ads, emotional storytelling
        Focus = memory + emotion

2) Local Advertising
        Target specific geographic area
        Example: Nearby stores, promotions
        Focus = conversion (short-term sales)

3) International Advertising
        Same product across countries
        Problem: Cultural differences, Language & Meaning mismatch

   
4) Corporate Advertising
        Promote company image, not product
        Example: Petronas ads, Tenaga Nasional Berhad
        Goal = trust + reputation + legitimacy

5) Public Service Ads (PSA)
        Social causes (not profit)
        Example: Anti-smoking campaigns
        Goal = behavior change without financial incentive

#### Types of Reminder Advertising
Continuous Ads
    
    Ads shown regularly (every month)

Seasonal Reminder

    Shown during special times
    Festive ads (Hari Raya, Christmas)

Emotional Reminder

    Focus on feelings, not product features
    Family, happiness, friendship

Reminder ads don't try to convince you again, they rely on:

    Habit
    Familiarity
    Emotional connection

This is called:

    "Mere Exposure Effect"

The more you see something, the more you like it.

### d ) Functions
1) Brand Awareness
        People know your brand exists, Probability of recall increases.
        Awareness = memory encoding in brain
2) Brand Image
        What people think/feel about brand
        Example:
            Apple → premium, smart
            Nike → performance, motivation
        Image = perception layer
3) Brand Equity
        Added value from brand perception
        Example:
            Same product, different price due to brand
        Equity = pricing power + psychological advantage
4) Communication & Interaction
        Connect brand with audience
        Modern ads = two-way system (comments, shares, engagement)

##### All 4 functions form a pipeline:

    Awareness → Image → Equity → Profit

## 2: AIDA Model  
### a ) Attention 
Break through noise and capture the audience's focus instantly.  
Psychology Behind It:

    Humans have limited attention span (especially online).
    Your brain filters information: only novel, emotional, or surprising things get noticed.

Techniques:
    
    Bright colors / bold visuals
    Shocking or surprising statements
    Humor 
    Influencers / celebrities
    Questions ("Are you tired of…?")

This stage is about interruption marketing, you are interrupting what people are doing (scrolling, watching). If you fail here, the rest doesn't matter.  

### b ) Interest
Keep the audience engaged and build curiosity. Psychology Behind It:
Now the brain asks:
        
    "Is this relevant to me?"

People only continue if they see:

    Personal benefit
    Relevance to their life

Techniques:

    Show a problem they relate to
    Tell a short story 
    Highlight key features/benefits
    Use relatable situations

    
Shift from attention-grabbing to value-delivering. This is where target audience knowledge is critical. 

### c ) Desire
Turning logic into want, How ads create desire:

    Show outcomes (before/after)
    Social proof (others use it)
    Status signaling (luxury, success)
    Emotional triggers (fear of missing out, insecurity, aspiration)
    Connection to Elaboration Likelihood Model

Two paths:

    Central route (logic-based)
    "This product is better because…"
    Peripheral route (emotion-based)
    "This feels right"
    Influencers, aesthetics, vibes

Desire is rarely rational—it's constructed emotionally, then justified logically.

### Action
Converting intention into behavior. This is where many ads fail. How ads trigger action:

    Clear CTA ("Buy now")
    Urgency ("Limited time")
    Scarcity ("Only 3 left")
    Reduced friction (easy checkout)

Psychology:

    Humans procrastinate
    Need a push to act now

Without friction reduction, desire dies before action.

## 3: Advertising: Communication Process
### a ) Sender (Strategic Intelligence Layer)
Not just "company", It includes:

    Brand positioning
    Market research
    Consumer psychology models
    Data (behavior, preferences)

Example, Nike doesn't just "send ads", They encode:

    Identity (performance, winning)
    Emotion (motivation, struggle)
    Social meaning (status, identity)
Sender = decision engine, not just a brand  

### b ) Message (Encoded persuation) 
Message is NOT just information,  It is:

    Symbols (images, words, colors)
    Emotional triggers
    Cognitive biases

Types of encoding:

    Rational → facts, features
    Emotional → fear, happiness, pride
    Social → belonging, status
Example, Apple ads:

    Minimal words
    Clean visuals
    → encode: premium + simplicity + intelligence
Message = compressed influence

### c ) Channel (Distribution + Algorithm)
Not just "TV or Instagram", Each channel has:

    Algorithm (what gets shown)
    Format constraints (video length, style)
    Audience behavior patterns
Example, TikTok:

    Short attention span
    Algorithm rewards engagement
    Hook needed in 1–3 seconds
Channel = filter + amplifier  
If you don't understand the algorithm → your message dies.
### d ) Receiver 
Old theory assumes that Receiver just "receives", but in reality, receiver actively:

    Filters information
    Interprets based on beliefs
    Ignores most ads
This is called:

    Selective Attention
    Selective Perception
    Selective Retention
Example:
    
    A gamer sees gaming ads
    A non-gamer ignores them instantly
Receiver = decision-maker, not target

### e ) Decoding (Where ads win or die)
This is the most underestimated part, What matters is not what you say but what the audience understands. Mismatch happens when:

    Message unclear
    Cultural misunderstanding
    Wrong tone
Example:

    Luxury brand uses humor → may reduce perceived prestige
Encoding != Decoding  
(You control encoding, not decoding)

### f ) Feedback (Data Layer)
Feedback is not just "sales", Modern feedback includes:

    Click-through rate (CTR)
    Watch time
    Conversion rate
    Engagement (likes, shares)
Platforms like Google Ads and Meta track everything.   
Feedback = signal for optimization loop

### g ) Noise (Enemy)
Noise = anything that disrupts communication.  
Types:

1) External Noise (Too many ads) / Distractions (phone notifications)
2) Psychological Noise (Mood, Bias & Prior beliefs)
3) Semantic Noise (Confusing message, Complex wording)
   
Example:  
User scrolling fast, skips your ad in 0.5 seconds

### h ) Creativity (Weapon)
Creativity solves noise, It does 3 things:

    (1) Pattern Interrupt (Breaks expectation)
        Example: Unexpected visuals, strong hook

    (2) Emotional Activation
        Emotion = memory
        Example: Fear, humor, inspiration

    (3) Cognitive Compression
        Simple message = faster understanding
        Example: "Just Do It" (Nike)

Creativity = attention engineering

#### Full System :
    Data → Strategy → Encoding → Channel Algorithm → Human Psychology → Behavior → Feedback → Optimization
<br>

## 4: The 5 main players
### a ) Advertisers
The company paying for ads. Examples:

    Procter & Gamble
    Unilever
    PepsiCo
Role:

    Decide budget
    Choose audience
    Control strategy
Advertiser = capital allocator

### b ) Advertising Agencies
Create and manage ads. Examples:

    Leo Burnett
Role:

    Strategy
    Creativity
    Media planning
Agency = execution + optimization engine

### c ) Media
Platform that delivers ads. Examples:

    TV, radio, social media
Media = distribution system + algorithm

### d ) Vendors
Support services, Examples:

    Designers
    Photographers
    Freelancers
Vendors = production layer

### e ) Audience
Target market

    Buying != consuming
    Example: Parents buy cereal, kids influence decision
Audience = decision network (A purchase decision is not made by one person alone—it's influenced by multiple people, factors, and interactions connected together)(not individual)

#### Full system :
    Advertiser (capital)
    → Agency (strategy + creative)
    → Media (distribution + algorithm)
    → Audience (psychology + behavior)
    → Feedback (data)
    → Optimization

Advertising is a system of influence operating across economics, psychology, and media technology.
